## DataFrame이란?

> - **DataFrame**이란 컴퓨터로 정보를 정리하고 저장하는 도구
> - Pandas에는 대표적으로 Series와 DataFrame이라는 데이터 오브젝트가 존재


### Series의 구조
> - Series는 값과 인덱스로 구성된, 1차원 자료구조
> - Series는 1차원 배열로 볼 수 있음
> - Series를 구성하는 값들은 모두 같은 데이터타입이어야 함


### DataFrame의 기본구조

> - **행**과 **얄**로 이루어진 2차원 배열 구조
> - 각 행과 열은 데이터 프레임 내에서 데이터의 위치를 나타냄
> - 시리즈가 모인 것이 하나의 데이터 프레임
---
## 데이터 확인, 추출, 조작

### 데이터 확인
1. `df.head(n)`, `df.tail(n)`을 통해, 앞뒤 일부 데이터를 미리 확인하여 구조, 값 확인
2. 전체 데이터에 대한 정보 확인: 
    - `df.shape` : 데이터의 행과 열 크기 확인
    - `df.info()` : 데이터 타입 및 결측치 정보 확인
    - `df.describe()` : 데이터의 기초 통계량 확인
    - `df.isnull()` : 결측치 여부 확인
    - `df.dtypes` : 데이터 타입 확인
3. 고유값 및 개수 확인
    - `df.count()` : 결측치 개수 제외한 행 수 반환
    - `df[열 이름].value_counts` : 특정 열에 포함된 값의 빈도수 반환
    - `df[열 이름].unique()` : 특정 열에 포함된 값을 종류별로 1개씩 반환

### 데이터 추출
1. 인덱싱 & 슬라이싱
    - `df[1]` -> 단일 항목 지정
    - `df[1:3]` -> 범위내 여러 항목 추출
2. 행단위 추출
    - `loc`
        > - 인자로 행과 열의 이름이 들어옴
        > - `df.loc[행이름]` , `df.loc[행이름, 열이름]`, `df.loc[:,열이름]`
    - `iloc`
        > - 인자로 정수형 인덱스 번호가 옴
        > - 나머지는 `loc`과 동일
3. 열단위 추출
    - 열단위로만 데이터 추출하고 싶으면, 열 이름으로 인덱싱하는 것이 직관적
    - `df[열이름]`
4. 불리언 인덱싱
    - True 또는 False의 배열을 통해 데이터를 필터링하는 연산
    - `df[조건문]`: 여기서 조건문은 여러가지를 논리 연산자로 묶어서도 사용 가능
5. 인덱스 설정
    - 인덱스를 원하는대로 변경 가능
    - `df.set_index(열 이름)` : 기존의 열을 인덱스로 설정
    - `df.reset_index()` : 인덱스를 리셋하고 기존 인덱스를 추가
    - `df.reindex(인덱스 배열)` : 인덱스를 재배열, 새로운 인덱스 생성도 가능

### 데이터 조작
1. 데이터 추가, 삭제
    - `df['새로운 열 이름] = 리스트`: 특정 데이터 셋ㅅ을 추가
    - `df.drop(labels, axis)` : 열 삭제
2. 데이터 정렬
    - `df.sort_index()` : 인덱스를 기준으로 정렬(기본값은 오름차순)
    - `df.sort_values(by="열이름")` : 특정 열의 값을 기준으로 정렬
3. 데이터 변환
    - `df[열이름] = df[열 이름].astype(바꿀 자료형)` : 자료형 변환
    - `df.replace(바꾸고 싶은 값, 바꿀 값)` : 데이터의 값 변환
    - `df.transpose()`,`df.T` :  행과 열 변환
4. 데이터프레임 병합
    - `df = pd.concat([df1,df2], axis = 0 or 1)` : 두개의 데이터 프레임을 합침 axis에 따라 행, 열방향 결합 달라짐
    - `df = pd.merge(df1,df2, how = 'inner/outer', on = '~~')` : on을 통해 기준 열 지정, inner은 교집합, outer은 합집합

## 집계함수
### 집계함수란?
- 여러행으로부터 하나의 결괏값을 반환하는 함수
- 대표적인 집계함수:
    - `df['컬럼 명'].mean()` : 평균
    - `df['컬럼 명'].median()` : 중앙값
    - `df['컬럼 명'].std()` : 표준편차
    - `df['컬럼 명'].sum()` : 합계
    - `df['컬럼 명'].max()` : 최대값
    - `df['컬럼 명'].min()` : 최소값


### 그룹화
- 집계함수는 모든 행을 대상으로 진행 하지만, 일부 행을 대상으로 집계할때 더 유용한 상황 존재
- 보통 `groupby`와 집계함수를 같이 사용
- `df.groupby('column명1')['column명2'].집계함수` : Series 형태로 결과 반환
- `df.groupby('column명1')[['column명2']].집계함수` : DataFrame 형태로 결과 반환
