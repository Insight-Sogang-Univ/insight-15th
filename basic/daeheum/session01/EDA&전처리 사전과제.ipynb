## EDA란?

> - 도표, 그래프, 통계등 다양한 각도에서 수집한 데이터를 여러가지 방식으로 파악하는 과정
> - 데이터의 특성, 패턴, 관계를 파악하고 시각화하면서, 후속 분석에 대한 방향성 제시
> - 시각화를 통해 패턴 발견, 데이터의 특이성 확인, 시각적 표현을 통해 가설 검증
---
### EDA 대상 (일변량/다변량)

#### 1.일변량
> - 일변량: EDA를 통해 한개의 변수를 파악
> - 데이터를 설명하고 그 안에 존재하는 패턴을 찾는 것이 주 목적

#### 2.다변량
> - 다변량: EDA를 통해 변수가 여러 변수들 간의 관계를 보는 것
> - 동시에 확인하기 전에 개별 데이터 먼저 파악

### EDA 종류(시각화/비시각화)

> - 시각적으로 파악할 것인지, 수치적으로 파악할 것인지

#### 1.Graphic(시각화)
> - 차트 혹은 그림을 이용하여 데이터를 확인하는 방법

#### 2.Non-Graphic(비시각화)
> - 그래픽적인 요소를 사용하지 않고 Summary Statistics를 통해 데이터를 확인하는 방법

#### 3.EDA 유형
> - Uni-Non Graphic, Uni-Graphic, Multi-Non Graphic, Multi-Graphic

## 데이터 확인

### 데이터 읽기/쓰기
> - `pd.read_csv` 를 통해 데이터를 읽어옴
> - 이때 파일경로에는 절대경로와 상대경로가 존재
```
pd.read_csv()
pd.read_exel()
```
> - csv파일인지, exel파일인지에 따라 다르게 읽어옴

### 데이터셋 파악하기

> - `df.head()`: 데이터 프레임 상위 n개 데이터 보여줌

#### 데이터의 종류
> - ***범주형***: 몇 개의 범주로 나누어진 데이터
    > - 명목형: 성별, 혈액형 등 순서와 상관없이 이름만 의미를 부여할 수 있는 데이터
    > - 순서형: 범주형 데이터 중 어떤 기준에 따라 순서에 의미를 부여할 수 있는 경우
> - ***수치형***: 숫자로 표현되는 데이터
    > - 이산형: 이산적인 값으로, 정수 단위로 떨어져 셀 수 있는 데이터
    > - 연속형: 연속적인 값을 갖는 데이터 (ex) 체중
> - `df.info()`: 데이터 형식 확인 -> 변수 데이터의 자료형 확인

## 전처리

> - 데이터 전처리: 데이터를 분석에 적합한 형태로 만드는 과정
> - 데이터 확인 -> 결측값 처리 -> 이상값 처리 -> 변수 가공

### 결측값 처리

> - 결측값 = 데이터 수집 과정에서 측정되지 않거나 누락된 데이터(Nan)

#### 결측값 확인

> - `info()`: 각 열에 속하는 NaN이 아닌 값의 개수를 보여줌
> - `value_counts(dropna = False)` : 각 열의 결측값을 포함한 전체 데이터 확인
> - `isnull()` : 누락 데이터면 True, 유효한 데이터면 False 반환
> - `notnull()` : 유효 데이터면 True, 누락 데이터면 False 반환
> - `replace()` : 결측값이 NaN이 아니라 

#### 결측값 처리

> - `dropna()` : DataFrame 내의 결측값이 포함된 레이블을 제거
> - 결측값이 무작위로 발생한 것이 아닌데 관측치를 삭제한 데이터를 사용할 경우, 왜곡된 모델이 생성될 수 있음
> - 따라서 `fillna()` : 결측값을 평균값 최빈값 등의 대푯값을 사용하여 대체

### 이상치 처리

> - 이상치 = 관측된 데이터의 범위에서 많이 벗어난 값

#### 이상치 확인

> - 통계(`describe()`), 시각화(BoxPlot), Z-score(데이터를 평균과 표준편차로 정규화), Tukey Fences를 통해 확인

#### 이상치 제거 및 처리

> - 전체 삭제
> - 다른 값으로 대체: 관측치를 평균으로 대체하거나, 이상값을 예측한 후 해당 값으로 대체
> - 변수화: 자연 발생한 이상값의 경우 삭제하지 않고, 이상값에 대해 파악
> - 리샘플링: 자연발생한 이상값을 분리해서 따로 모델을 만듬


### 피처 엔지니어링 (변수 가공)

> - 해결하고자하는 문제를 컴퓨터가 잘 이해할 수 있도록 변수들의 형태를 변형하거나 적절하게 처리
> - 기존의 데이터에 정보를 추가하는 과정 -> 데이터, 변수 추가 없이 기존 데이터를 유용하게 이용

#### 레이블 인코딩 vs 원핫인코딩
> - Label Encoding: 범주형 변수를 0~N-1까지의 숫자로 변환
> - One-Hot Encoding: 범주형 변수를 0과 1로 변환 -> `pd.get_dummies(df, columns=[컬럼리스트])`

#### 구간화
> - 연속 데이터를 그대로 사용하기 보다는 일정한 구간으로 나눠서 분석

#### 파생변수 생성
> - 기존의 피처를 다른 피처로 변환하여 변수 추가

#### 스케일링
> - 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업(정규화)

## 데이터 시각화

> - 데이터에 대한 이해를 돕기 위해 그림, 도형 등 그래픽 요소를 이용해 데이터를 묘사하고 표현하는 과정

> - 대표적으로 Matplotlib 와 Seaborn 사용

### 파라미터
> - 매개변수란 프로그래밍된 함수의 입력값을 의미함

### 주요 그래프 종류들

1. boxplot -> 사분위수와 이상치를 보여주는 그래프
2. countplot, histplot
    - coutplot: 범주형 변수의 빈도수 확인
    - histplot -> 도수 분포표를 나타낸 그래프
3. displot, kdeplot -> 히스토그램을 연속적으로 곡선으로 연결한 그래프
4. barplot, pointplot
    - barplot: 범주형 데이터 값 x에 따른 수치형 데이터 값의 평균 값
    - pointplot: 막대 그래프와 모양만 다르고, barplot과 동일한 정보
5. scatterplot, regplot
    - scatterplot: 두 연속형 변수 간의 관계 시각화
    - regplot: 두개의 연속 변수 사이의 산점도를 그리고 회귀선도 함께 나타냄
6. catplot: 수치형 데이터와 범주형 데이터의 관계를 볼때 주로 사용
7. pieplot: 데이터의 부분과 전체 간의 비율을 표현
8. heatmap: 변수간 상관계수를 직접적으로 볼 수 있는 그래프
    `` sns.heatmap(df.corr(), annot = True)
    ``
9. violinplot: 박스 플롯과 커널 밀도 추정 함수 그래프를 합쳐 놓은 것
    `` sns.violinplot(x='class',y='age', data = data)
    ``
10. pairplot: 여러 그래프를 한번에 설정
