### 3. 데이터 확인, 추출, 조작

**(1) 데이터 확인 방법**

방대한 크기의 데이터를 다루는 경우가 많기에 일부만 미리 확인하거나 전체적인 특징을 요약해주는 함수 사용

ex) head, tail, shape, info, describe, isnull, dtypes, count, value counts, unique

**(2) 데이터 추출 방법**

<인덱싱 및 슬라이싱>
* 인덱싱 : 단일 항목을 지목
* 슬라이싱 : 범위를 지정해 여러 항목을 추출할 때 사용

<행 단위 추출>

* loc : 인덱스 이름 기준으로 행 성택
* iloc : 정수형 위치를 사용할 때 필요

<열 단위 추출> 

* 컬럼명

<볼리언 인덱싱>

* True of False 배열을 통해 데이터를 필터링 하는 연산
* 인덱스나 컬럼명이 아니라 특정 조건을 만족하는 데이터 추출할 때 사용

<인덱스 설정>

* 인덱스는 0부터 시작하는 정수형 인덱스를 기본값으로 가지지만, 원하는 대로 변경 가능

**(3) 데이터 조작**

* 데이터 추가, 삭제 : 열을 추가 및 삭제
* 데이터 정렬 : 새로운 방식으로 정렬
* 데이터 변환 : 특정 컬럼 내 데이터의 자료형 변환
* 데이터프레임 병합 : concat/merge

### 4. 집계함수 정의 및 사용방법

**정의 : 여러 행으로부터 하나의 결괏값을 반환하는 함수**

* 각 컬럼의 평균, 표준편차, 합계 등 연산을 통해 대표값을 구할 때 주료 사용
* 결측치가 있을 경우 결측치는 무시하고 계산

**대표적인 집계함수**

* `df['컬럼 명'].mean()` : 평균
* `df['컬럼 명'].median()` : 중앙값
* `df['컬럼 명'].std()` : 표준편차
* `df['컬럼 명'].sum()` : 합계
* `df['컬럼 명'].max()` : 최대값
* `df['컬럼 명'].min()` : 최소값
