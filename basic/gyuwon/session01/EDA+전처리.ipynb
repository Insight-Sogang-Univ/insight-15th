### **1. EDA(Exploratory Data Analysis)**

**(1) EDA 란?**

>탐색적 데이터 분석 과정 : 다양한 각도에서(도표, 그래프, 통계) 수집한 데이터의 특징을 파악하는 모든 과정

- 시각화 같은 도구를 통해서 패턴 발견
- 데이터의 특이성 확안
- 통계와 그래프(혹은 시각적 표현)으로 가설을 검정하는 과정


*을 통해 주어진 데이터에 대해 알아보는 것*

---

**(2) EDA 대상**

> 하나의 변수를 살펴볼 것인가, 여러 변수 간의 관계를 살펴볼 것인가?

* **Unvariate(일변량)**

    EDA 를 통해 한 번에 파악하려는 변수가 한 개.
    데이터를 설명하고 그 안에 존재하는 패턴을 찾는 것이 주요 목적

* **Multi-variate(다변량)**

    EDA 를 통해 한 번에 파악하려는 변수가 여러 개
    여러 변수들 간의 관계를 보는 것이 주요 목적
    변수를 동시에 확인하기 전에 개별 데이터를 먼저 파악하는 것이 오류에 대처하기 용이함

---

**(3) EDA 종류**

* **Graphic(시각화)**

> 차트 혹은 그림 등을 이용하여 데이터를 확인하는 방법 -> 대략적인 형태

* **Non-Graphic(비시각화)**

> 그래픽적인     요소를 사용하지 않고 주로 Summary Statistics 를 통해 데이터를 확인하는 방법 -> 정확한 값

### **3. 전처리**

#### **(1) 결측값**

**결측값 : 데이터 수집 과정에서 측정되지 않거나 누락된 데이터**

    결측값이 있는 상태로 데이터 분석 모델을 만들면 변수 간의 관계가 왜곡될 수 있어 모델의 정확성이 떨어져 시각적 표현에도 문제를 일으킴

---

> ##### ***결측값 확인***

* info()
* value_counts(dropna=False)
* isnull() -> df.isnull().sum
* notnull()
* replace()

___

> ##### ***결측값 처리***

##### **삭제** : 데이터가 있는 행 또는 열을 삭제

```python
DataFrame.dropna(axis=0/1, how='any'/'all', subset=[col1, col2, ...], inplace=True/False)
```

    ‼️주의할 점‼️

***결측값이 아닌 관측치를 삭제하면 정보가 손실 돼 왜곡된 데이터 분석 모델이 생성될 수 있음!***

##### **대체 : 결측값을 다른 값으로 대체**

* 일괄 대체

```python
# 평균값으로 일괄 대체
a = df["열1"].mean(axis=0)
df["열1"].fillna(a, inplace=True) # 원본객체를 변경할시에 inplace=True
```

* 유사 유형 대체

```python
# 범주별로 그룹화하여 평균값 계산
category_means = df.groupby('Category')['Value'].mean()

# 결측값을 범주의 평균값으로 대체
df['Value'] = df.groupby('Category')['Value'].apply(lambda x: x.fillna(x.mean()))
```

#### **(2) 이상치**

**이상치 : 관측된 데이터의 범위에서 많이 벗어난 값**

> ##### ***이상치 확인***

* **통계** 를 통한 확인 :

```python
describe()
```

* **시각화**를 통해 확인 : BoxPlot

* **Z-score** 를 통해 확인 : 평균과 표준편차로 정규화 하여 평균으로부터 얼마나 떨어져있는가?

* **Turkey Fences** 를 통해 확인 : **사분위 범위(IQR, Interquartile Range)** 를 기반으로, 두 가지 경우에 이상치라고 판단
    
    - Q1 - (1.5 * IQR) 미만
    - Q3 + (1.5 * IQR) 초과

> ##### ***이상치 제거 및 처리***

* **전체 삭제** : Human error 에 의한 발생한 경우

* **다른 값으로 대체** : Human error 에 의해 발생했지만 절대적인 관측치의 숫자가 작은 경우

* **변수화** : 이상값이 자연 발생한 경우

* **리샘플링** : 자연발생한 이상값을 처리할 때 해당 이상값을 분리해서 모델 만드는 방법

### **4. 피처 엔지니어링(변수 가공)**

> ***해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 도메인 지식과 기존의 변수를 사용해서 기존의 데이터에 정보를 추가하는 일련의 과정***

##### ✅ **피처 엔지니어링 방식**

* **레이블인코딩, 원핫인코딩** : 문자열 값들을 숫자형으로 변경하는 작업
```python
data_dummies=pd.get_dummies(data['Survived'], drop_first=True, dtype=int)
```
---
* **구간화(Binning)** : 연속적인 값을 일정한 구간(Bin) 으로 구분
```python
count, bin_dividers = np.histogram(df["카테고리"], bins=3) #3개의 구간으로 나눔 
bin_names = ["상", "중", "하"]
df["new_bins"] = pd.cut(x=df["카테고리"],
bins = bin_dividers,
labels = bin_names, # bin 이름
include_lowest = True) # 첫 경계값 포함
```
---
* **파생변수 생성** : 기존의 피처를 변환해 새로운 피처 생성
    - 단위 변환, 표현 형식 변환, 요약 통계량 변환, 정보 추출, 변수 결합, 조건문 이용
---
* **스케일링** : 사로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업

##### **(3) 다양한 그래프**

* ***BowPlot(상자 수염 그림) : 사분위수와 이상치를 보여주는 그래프***

* ***CountPlot/Hisplot : 범주형 변수의 빈도수를 확인하는 그래프/도수분포표를 그래프로 나타내 수치형 변수의 구간별 빈도수 확인하는 그래프***

* ***DisPlot, KdePlot : 히스토그램을 연속적으로 곡선으로 연결한 그래프***

* ***BarPlot/PointPlot : 범주형 데이터 값 x 에 따른 수치형 데이터 값 y 의 평균값을 제공/막대 그래프와 모양을 제외하고 동일***

* ***ScatterPlot/RegPlot : 두 연속형 변수 간의 관계를 시각화하기 위해 사용되는 그래프/두개의 연속 변수 사이의 산점도를 그리고 회귀선을 함께 나타내는 그래프***

* ***CatPlot : 수치형 데이터와 범주형 데이터의 관계 확인***

* ***PiePlot : 데이터의 부분과 전체 간의 비율을 표현하는 그래프***

* ***HeatMap : 변수간 상관계수를 직관적으로 볼 수 있는 그래프***

* ***ViolinPlot : 박스 플롯과 커널밀도추정 함수 그래프를 합쳐 놓은 그래프***

* ***PairPlot : 여러 변수간의 산점도를 한 번에 보여주는 그래프***
