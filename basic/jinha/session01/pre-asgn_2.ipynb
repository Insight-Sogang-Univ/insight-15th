## 데이터 탐색 (EDA)

- Exploratory Data Analysis의 약자로, 탐색적 데이터 분석 과정을 일컫는 말
- 데이터 분석(모델링)을 위해 데이터를 여러 가지 방식으로 파악하는 모든 과정
- 다양한 각도에서 (도표, 그래프, 통계) 수집한 데이터의 특징을 파악
- 데이터의 특성, 패턴, 관계를 파악하고 시각화 하면서 데이터를 이해하고 전처리, 모델링 등의 후속 분석에 대한 방향성을 제시하는 단계

## 데이터셋 파악하기

1. 데이터 프레임 보기
2. 데이터 변수 확인
- 변주 정의, 유형, 자료형 확인
3. 데이터 분포 확인
- 산포도, 박스 플롯, 히스토그램 등의 그래프를 통해 확인 (단변수 분석)

## 결측값 & 이상치 처리

### 결측값 처리

1. 결측값 확인
2. 결측값 처리

### 이상치 처리

1. 이상치 확인
- 통계를 통해 확인
- 시각화를 통해 확인

2. 이상치 제거 및 처리
- 전체 삭제

- 변수화
- 리샘플링

## 피처 엔지니어링 (변수 가공)

- 해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 피처(변수)들의 형태를 변형하거나 적절하게 처리하는 과정

### 레이블인코딩(Label Encoding) vs 원핫인코딩(One-hot Encoding)
- Label Encoding : 범주형 변수를 0부터 N-1까지의 숫자로 변환
- One-Hot Encoding : 범주형 변수를 이진 벡터(0과 1)로 변환

### 구간화 (Binning)
- 연속 데이터를 일정한 구간으로 나눠서 분석

### 파생변수 생성
- 기존의 피처를 다른 피처로 변환(재정의)하여 변수를 추가
- 단위 변환, 표현형식 변환, 요약 통계량 변환, 정보 추출, 변수 결합, 조건문 이용 등의 방법

### 스케일링
- 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업

## 데이터 시각화

- 데이터에 대한 이해를 돕기 위해 그림, 도형 등 그래픽 요소들을 이용해 데이터를 묘사하고 표현하는 과정
