## <span style="background-color:#fae0f0; color:#c0156d; padding:4px; border-radius:5px;"> 2. 데이터프레임(DataFrame)의 구성 </span>

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 2.1 개념 </span>
- 컴퓨터로 정보를 정리 & 저장하는 도구
- 판다스에서 사용하는 주요 데이터 구조
    - <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> **Series** </span> : 1차원 배열의 형태, 인덱스(행)에 의해 데이터가 저장됨
    - <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> **DataFrame** </span> : 2차원 배열의 형태, 인덱스(행), 컬럼(열)에 의해 표 형태로 데이터가 저장됨

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 2.2 Index & Column </span>
1. **개념**
    - **행** : **가로**로 나아가는 방향 <br>
    - **열** : **세로**로 나아가는 방향 <br>
2. **특징**
    - 데이터프레임의 기본구조 = 행(Index) + 열(Column) <br>
    - 행, 열로 데이터의 위치를 나타낼 수 있음 <br>

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 2.3 시리즈(Series) </span>
1. **개념**<br><br>
    - **값, 인덱스**로 구성된 가장 간단한 **1차원 구조** -> 행(Index)가 있는 1차원 배열(array) <br>
2. **특징 및 조건**<br><br>
    - 인덱스 레이블 지정 X -> 디폴트 정수형 인덱스 자동 부여 (0부터 시작) <br>
    - 데이터프레임에서 특정 **컬럼 1개만 추출 가능**
    - 구성 값은 **동일한 데이터 타입**이어야 함 <br>
        - 🙋‍♀️ <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> 시리즈가 여러 데이터 타입으로 구성된다면? </span>
            - 전체 데이터는 **문자열**로 처리됨
            - **산술 연산 불가능** -> 데이터 타입을 **숫자로 변환**하는 과정 필요 (`astype(int)`, `pd.to_numeric()`)
3. **시리즈 만드는 방법**<br><br>
    - **리스트 생성**
    > ```python
    > series = ['사과', '포도', '복숭아', '배', '감']
    > ```
    - **시리즈 객체 생성** `pd.Series()` 활용 (인덱스 지정 X)
    > ```python
    > sr = pd.Series(series)
    > print(sr)
    > ```
    - **인덱스 추가** (인덱스 직접 지정 시)
    > ```python
    > index = [1,2,3,4,5,6]
    > sr = pd.Series(series, index = index)
    >
    > print(sr)
    > ```

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 2.4 데이터프레임(DataFrame) </span>
1. **개념**<br><br>
    - **행, 열**로 만들어진 **2차원 배열 구조** <br>
2. **특징 및 조건**<br><br>
    - 여러 개의 시리즈 -> 데이터프레임 <br>
    - 컬럼(시리즈) 간 데이터 타입은 달라도 무방
3. **데이터프레임 만드는 방법**<br><br>
    - **딕셔너리 생성**
    > ```python
    > data = {
    >   '카테고리1' : ['데이터1', '데이터2', '데이터3'],
    >   '카테고리2' : ['데이터4', '데이터5', '데이터6'],
    >   '카테고리3' : ['데이터7', '데이터8', '데이터9']
    > }
    > ```
    - **데이터프레임 구성** `pd.DataFrame()` 활용
    > ```python
    > df = pd.DataFrame(data)
    > ```
    - **인덱스, 컬럼 직접 지정** (기본값은 0부터 시작)
    > ```python
    > index = [1,2,3,4,5,6]
    > df = pd.Data(data, index = index)
    >
    > print(df)
    > ```

## <span style="background-color:#fae0f0; color:#c0156d; padding:4px; border-radius:5px;"> 3. 데이터 확인&추출&조작  </span>

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 3.1 데이터 확인 </span>
> 데이터 분석 전,  <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> **데이터 크기,타입, 값의 분포, 경향 등을 파악**하는 단계가 필요 </span> <br>
> 따라서, 아래와 같은 함수를 사용해, 방대한 데이터를 빠르게 파악할 수 있음

1. **head(), tail()** <br><br>
    > 앞/뒤의 일부 데이터를 미리 확인 -> <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> 데이터의 개괄적인 **구조, 값** 파악 </span>
    - `df.head(n)` : 앞 n개의 행 반환 (기본값 n=5)
    - `df.tail(n)` : 뒤 n개의 행 반환 (기본값 n=5)

2. **데이터 정보 확인** <br><br>
    > 데이터프레임의 <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> 행/열 개수, 열 별 데이터타입, 결측치 개수 등 확인 </span>
    - `df.shape` : (행 개수, 열 개수) 반환
    - `df.info()` : 여러 정보들 종합 반환 (행개수, 열개수, 데이터타입 등)
    - `df.describe()` : 수치형 열 별 기본적인 통계값 반환
    - `df.insull()` : 결측치(누락값) 유무 여부 반환 
    - `df.dtypes` : 열 별 데이터타입 반환 <br><br>
        - 💡 **판다스 데이터타입 종류** <br>

            | 판다스 | 파이썬 | 비고 |
            | --- | --- | --- |
            | int64 | int | 정수형 |
            | float64 | float | 실수형 |
            | object | string | 문자열 |
            | datetime64, timedelta64  | X(datetime 라이브러리 활용) | 시간 |

3. **고유값 & 개수 확인** <br><br>
    > <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> 데이터 전체 개수, 고유값 파악 </span>
    - `df.count()` : 데이터의 행 수 반환 (결측치 개수 제외)
    - `df[column 이름].value_counts()`: 특정 열에 포함된 값들의 빈도수 반환
    - `df[열 이름].unique()` : 특정 열에 포함된 값들을 종류별로 1개씩만 반환

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 3.2 데이터 추출 </span>
> 데이터 분석 시,  <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> **분석에 필요한 부분만 골라내 가공**하는 능력이 중요 </span> <br>
> 따라서, 아래와 같은 함수를 사용해, 방대한 데이터를 빠르게 추출할 수 있음

1. **인덱싱&슬라이싱** <br><br>
    > index, column을 사용해 데이터를 추출하는 것
    > 인덱싱은 차원을 하나 축소. 슬라이싱은 차원 변경 X
    - 인덱싱 : `df[컬럼 이름]`(단일 항목 지목 시 사용)
    - 슬라이싱 : `df[start:end]` (end 인덱스는 포함 X)

2. **행단위 추출** <br><br>
    - `loc` : 행과 열의 이름(label)으로 행 데이터 선택
        - 인덱싱 & 슬라이싱 모두 가능
        - 범위 : 시작 인덱스 ~ 끝 인덱스
        - `df.loc[행 이름]` : 행단위 추출
        - `df.loc[행 이름, 열 이름]` : 값단위 추출
        - `df.loc[:, 열 이름]` : 열단위 추출 (전체 행 중 컬럼이 '열 이름'인 데이터 추출)
    - `iloc` : 정수형 인덱스 번호로 행 데이터 선택
        - 방대한 데이터에서 loc보다 더 빠름
        - 범위 : 시작 인덱스 ~ 끝 인덱스 -1
        - `df.iloc[행 순서, 열 순서]` 
            - ex. `df.iloc[1:4, 1:]` : 인덱스 순서가 1~3번째, 컬럼 순서 1번째인 데이터 추출

3. **열단위 추출 : 컬럼명** <br><br>
    > 열 단위로만 데이터 추출 시 loc&iloc보다 더 직관적
    - `df[열 이름]` : 시리즈 반환
    - `df[[열이름]]` : 데이터프레임 자료형 유지
    - `df[[열 이름1, 열 이름2...]]` : 부분적인 데이터프레임 반환
    - `df[열 이름][행 이름]` : 특정 값 반환

4. **불리언 인덱싱** <br><br>
    > **특정 조건에 만족**하는 데이터 추출 시 사용
    - `df[조건문] `True or False 배열로 데이터를 **필터링**하는 연산
    - `df[(조건문1) 논리연산자 (조건문2)]` (논리연산자 : and, or)
    - `df.loc[조건문, 열 이름]` (iloc는 불가능)

5. **인덱스 설정** <br><br>
    > 인덱스 기본값 : 0부터 시작하는 정수형 인덱스 but 원하는 대로 변경 가능!<br>
    > ()안에 **inplace=True** 추가 시 df에 직접적으로 수정이 됨<br>
    > ()안에 **drop=True** 추가 시 기존 인덱스가 열에 추가되지 않고 사라짐 (디폴트 : F)<br>
    - `df.set_index(열 이름)` : 기존 열 중 하나를 인덱스로 설정
    - `df.reset_index()` : 인덱스 리셋, 기존 인덱스 열에 추가
    - `df.reindex(인덱스 배열)`
        - 행 인덱스 재배열 -> 인덱스 재배열에 따라 기존 행데이터도 재배열
        - 새로운 인덱스 생성 -> 값들에 결측치 기록
        - `df.index = [인덱스 배열]` : 인덱스 재설정 but 데이터는 고정

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 3.3 데이터 조작 </span>

1. **데이터 추가, 삭제** <br><br>
    > index, column을 사용해 데이터를 추출하는 것
    - `df['새로운 열 이름'] = 리스트` : 특정 리스트를 테이터 열로 추가
    - `df.drop(labels, axis)` : 특정 열을 삭제
        - axis(축) : n차원 배열을 구성하는 방향
            - axis=0 : 행 방향 연산
            - axis=1 : 열 방향 연산

2. **데이터 정렬** <br><br>
    - `df.sort_index()` : 인덱스를 기준으로 정렬
        - 기본값 : ascending=True (오름차순) 
    - `df.sort_values(by="열 이름")` : 특정 열의 값을 기준으로 정렬

3. **데이터 변환** <br><br>
    - **자료형 변환**
        - `df[열 이름] = df[열 이름].astype(바꿀 자료형)`
        - 참고 : `df.dtypes` : 열들의 데이터 타입 반환
    - **데이터 값 변환**
        - `df.replace(바꾸고 싶은 값, 바꿀 값)`
        - 기본값 : inplace=False
    - **행 <-> 열 전치**
        - `df.transpose()` = `df.T`
        - 기본값 : inplace=False
        - 바뀐 형태 저장 방법 : `df1 = df.transpose()`

4. **데이터프레임 병합** <br><br>
    - **concat** : 데이터 프레임을 행 or 열 방향으로 이어붙이기 (보통 동일한 구조의 DF)
        - `df = pd.concat([df1,df2], axis=0)` : 행방향 결합
        - `df = pd.concat([df1,df2], axis=1)` : 열방향 결합
    - **merge** : 공통 항목을 기준으로 병합
        - - `df = pd.merge(df1, df2, how = 'inner', on = '~~')` : 교집합(and) 연산
        -  `df = pd.merge(df1, df2, how = 'outer', on = '~~')`: 합집합(or) 연산
            - 겹치지 않는 부분은 NaN으로 반환

## <span style="background-color:#fae0f0; color:#c0156d; padding:4px; border-radius:5px;"> 4. 집계함수  </span>

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 4.1 집계함수 개념 </span>
- **여러 행으로부터 하나의 결괏값을 반환**하는 함수
- 연산으로 **대표값**을 구할 때 사용 <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> (각 컬럼의 평균, 표준편차, 합계 등) </span>
- 결측치는 무시

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 4.2 집계함수 사용 방법과 함수 종류 </span>
- **사용 예시**<br><br>
    > ```python
    > df[['열 이름1','열 이름2']].mean()
    > ```
- **대표적인 종류**<br><br>
    - `df['컬럼 이름'].mean()` : 평균
    - `df['컬럼 이름'].median()` : 중앙값
    - `df['컬럼 이름'].std()` : 표준편차
    - `df['컬럼 이름'].sum()` : 합계
    - `df['컬럼 이름'].max()` : 최대값
    - `df['컬럼 이름'].min()` : 최소값

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 4.3 그룹화 </span>
> 일부 행을 대상으로 집계할 때 활용 <br>
- <span style="background-color:#fff615; color:#000000; padding:0px; border-radius:5px;"> **groupby** 메소드 </span> : 특장 기준으로 그룹을 묶고, 연산 처리 (집계함수와 함께 사용)  <br><br>
    - `df.groupby('column1 이름')[’column2 이름'].집계함수`
    - column1으로 그룹화 -> column2에 집계함수 적용해 결과를 반환
