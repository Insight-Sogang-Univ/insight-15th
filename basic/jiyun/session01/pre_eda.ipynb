## <span style="background-color:#fae0f0; color:#c0156d; padding:4px; border-radius:5px;"> 2. EDA </span>

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 2.1 EDA 개념 </span>
> **E**xploratory **D**ata **A**nalysis (탐색적 데이터 분석 과정)
- 즉, 데이터 분석을 위해 <span style="background-color:#fffa7d; color:#000000; padding:0px; border-radius:5px;">**데이터를 여러 가지 방식으로 파악**하는 모든 과정</span>을 일컬음
     - 시각화를 통한 패턴 발견, 데이터의 특이성 확인, 통계/그래프를 활용한 가설 검정 등

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 2.2 EDA 대상  </span>
#### 1. Univariate(일변량)
- 한 번에 파악하려는 변수가 1개
- 목적 : 데이터 설명, 데이터 속 패턴 탐색

#### 2. Multi-variate(다변량)
- 한 번에 파악하려는 변수가 여러 개
- 목적 : 여러 변수 간 관계 파악
     - 주의 : 개별 데이터 파악이 선행되는 것이 오류 대처에 용이

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 2.3 EDA 종류  </span>
#### 1. Graphic(시각화)
- 차트, 그림 등을 통한 데이터 확인
- 장점 : 데이터를 한 눈에 파악 가능

#### 2. Non-Graphic(비시각화)
- Summary Statistics를 통한 데이터 확인
- 장점 : 정확한 값을 파악 가능

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 2.4 EDA 유형  </span>
> **EDA 유형 구분 요인**
> 
> - 데이터 변수 개수
> - 결과 파악 방법
> - 데이터의 유형

|             | **일변량(Uni)**                                            | **다변량(Multi)**                          |
|-------------|------------------------------------------------|--------------------------------|
| **시각화(Graphic)**     | **목적** : 데이터의 Distribution 확인 <br> **예시** : 파이차트, 막대그래프, 히스토그램, 박스플롯 | **목적** : 데이터를 전체적으로 파악 <br> **예시** : 모자이크플롯, 박스플롯, 평행좌표, 산점도|
| **비시각화(Non-Graphic)** | **목적** : 둘 이상의 변수 간 관계를 전체적으로 파악 <br> **예시** : 빈도표, 기술통계량| **목적** : 둘 이상의 변수 간 관계를 확인 <br> **예시** : 교차표, 상관계수|

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 4.2 결측값 처리 </span>
> **결측값** : 측정 X or 누락된 데이터 (NaN) <br>
> 결측값으로 인해 생길 수 있는 문제 : 변수 간 관계 왜곡, 모델의 정확성 감소, 시각적 표현에서의 오류

#### 1. 결측값 확인
- `NaN`, `?`, `0` 값의 결측값 확인
- 판다스 함수 활용 가능

    | 함수                       | 설명                                     |
    |----------------------------|------------------------------------------------|
    | info()                     | 요약 정보 출력 (non-null의 개수를 보여줌)     |
    | value_counts(dropna=False) | 결측치 포함한 전체 데이터 확인          |
    | **isnull().sum()**            | 빈칸(NaN) 개수만 총합으로 보기 (가장 많이 씀!) |
    | notnull()                  | 정상 데이터만 골라내기 (유효 데이터 = True)                       |
    | replace()                  | 수상한 값(?, 0)을 빈칸(NaN)으로 변환     |

#### 2. 결측값 처리
1. **<span style="background-color:#fffa7d; color:#000000; padding:0px; border-radius:5px;">삭제</span>** : `dropna()` 데이터가 있는 행, 열 삭제
- 결측값이 무작위로 발생한 경우에만 삭제. OR 정보가 손실돼, 왜곡된 모델 생성됨 <br>


    ```python
    DataFrame.dropna(axis=0/1, how='any'/'all', subset=[col1, col2, ...], inplace=True/False)
    ```
    <br>

    - `axis=0` vs `axis=1`
        - 0: 결측값이 있는 행 삭제 (관측값 삭제) (기본값)
        - 1 : 결측값이 있는 열 삭제 (변수 삭제)
    - `how = 'any'` vs `how='all'`
        - any : 하나라도 결측값 있으면 삭제 (기본값)
        - all : 모든 값이 결측값일 때만 삭제
    - `subset = ['column명']`
        - 특정 열 기준으로 삭제, `axis = 0` 일 때만 적용 가능
    - `inplace=False` vs `inplace=True`
        - F : 반영한 내용을 새로운 DataFrame으로 반환
        - T : 원본 DataFrame 자체를 변경해 반영

2. **<span style="background-color:#fffa7d; color:#000000; padding:0px; border-radius:5px;">대체</span>** : `fillna()` 결측값을 다른 값으로 대체
- 평균값, 최빈값 등의 대푯값을 활용 <br>

    - **<span style="background-color:#fffa7d; color:#000000; padding:0px; border-radius:5px;">일괄 대체</span>** : 모든 변수들의 평균값(대푯값)을 구해 일괄 대체
        ```python
        a = df["열1"].mean(axis=0)
        df["열1"].fillna(a, inplace=True)
        ```
    <br>

    - **<span style="background-color:#fffa7d; color:#000000; padding:0px; border-radius:5px;">유사 유형 대체</span>** : 유사한 범주에 따라 다른 값으로 대체
        ```python
        a = df["열1"].mean(axis=0)
        df["열1"].fillna(a, inplace=True)
        ```

        - **이웃한 값**으로 대체하는 방법도 있음
        ```python
        # NaN 직전 행에 있는 값
        df["열2"].fillna(method="ffill", inplace=True)

        # NaN 직후 행에 있는 값
        df["열3"].fillna(method="bfill", inplace=True).
        ```

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 4.3 이상치 처리 </span>
> 이상치 : 관측 데이터의 범위에서 많이 벗어난 값

#### 1. 이상치 확인
- 방법 1 : **통계** `describe()` <br><br>
- 방법 2 : **시각화**를 통해 확인 - **BoxPlot**
    ```python
    for i in df.describe().columns:
        df[[i]].boxplot()
        plt.show()
    ```
    <br>
- 방법 3 : **Z-score**을 통한 확인 <br>
    - 데이터를 평균(0), 표준편차(1)로 정규화해, 평균으로부터 얼마나 떨어져있는지 나타냄
    - 평균에 가까울 수록 0에 가깝고, 멀어질 수록 Z-score 증가
    - *Z-score > 2~3* : 이상치로 간주
    <br>

- 방법 4 : **Tukey Fences**를 통한 확인<br>
    - IQR은 데이터의 중간 50% 범위를 측정함 (하위, 상위 25% 제외)
    - 사분위 범위(IQR)를 기반으로 판단
        - Q1 - (1.5 * IQR) 미만 or Q1 - (1.5 * IQR) 초과
        - 보통은 1.5이나, 특성에 따라 3배 등으로 조절 가능

#### 2. 이상치 제거 및 처리
- 방법 1 : **전체 삭제**
    - Human error에 의해 발생한 경우 (단순 오타, 주관식 설문에서의 비현실적인 응답, 데이터 처리 과정에서의 오류 등)

- 방법 2 : **다른 값으로 대체**
    - 절대적인 관측치의 숫자가 작은 경우 : 삭제하면 관측치의 절대량이 작아져 신뢰성 문제 야기
        - 다른 값(평균 등)으로 대체
        - 결측값과 유사하게 다른 변수들을 사용해 예측 모델 만든 후, 이상값을 예측해 그 값으로 대체

- 방법 3 : **변수화**
    - 이상값이 자연발생한 경우 : 대체/삭제를 통해 모델 수립 시 현상에 대한 설명력 감소
        - 삭제/대체 전 찬찬히 이상값에 대해 파악하는 것이 중요
        - **이상값과 다른 값들을 구별지을 수 있는 변수**가 있다면, 이를 추가하여 이상값을 삭제하지 않고 모델에 포함할 수 있음 

- 방법 4 : **리샘플링**
    - 이상값이 자연발생한 경우 : 대체/삭제를 통해 모델 수립 시 현상에 대한 설명력 감소
        - 이상값을 분리해 **추가 모델**을 만들어 각각에 모델에 대한 설명을 달기
        - 독립 변수, 종속 변수가 모두 outlier인 경우 효과적!

#### 3. <span style="background-color:#fffa7d; color:#000000; padding:0px; border-radius:5px;">Tukey Fences</span>를 활용한 이상치 확인&제거 예시
```python
# 이상치 찾기 (Tukey Fences 이용)
def find_outlier_by_Tukey_DF(data, feature):
    q1, q3 = np.percentile(data[feature],[25,75])
    iqr = q3-q1
    
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    print(f'lower_bound는 {lower_bound.round(3)}, upper_bound는 {upper_bound.round(3)} 입니다')
    
    mask = data[(data[feature] > upper_bound) | (data[feature] < lower_bound)].index
    return mask.     

mask = find_outlier_by_Tukey_DF(df, 'BMI')
print(f'해당 열의 이상치는 {len(mask)}개 입니다')

# 이상치에 해당하는 행 df에서 제거
for i in range(len(mask)):
    df = df.drop(index=mask[i])
df = df.reset_index(drop=True)
```

### <span style="background-color:#e6fae0; color:#42c015; padding:4px; border-radius:5px;"> 4.4 피처 엔지니어링 </span>
> 컴퓨터가 잘 이해할 수 있도록, **feature(변수)의 형태를 적절히 변형&처리**하는 과정
- 도메인 지식 & 기존의 변수를 사용해 기존의 데이터에 정보를 추가하는 과정
- 데이터 전처리의 마지막 단계로, 새로운 데이터 or 변수 추가 없이 기존 데이터를 유용하게 만드는 방법

#### 1. 레이블 인코딩 vs 원핫 인코딩
> 문자열 값들을 기계가 이해할 수 있는 숫자형으로 변경하는 작업
1. Label Encoding
    - 범주형 변수를 0 ~ N-1까지의 숫자로 변환
2. One-Hot Encoding
    - 범주형 변수를 이진 벡터(0,1)로 변환 (표현하고 싶은 단어의 인덱스 = 1)
    - `pd.get_dummies(데이터프레임, columns=[컬럼 리스트])` : 명목변수를 0,1으로 원-핫 인코딩
    - 예시 코드
        ```python
        #pd.get_dummies()로 변수 data_dummies에 Survived 칼럼을 변환
        #drop_first=True : 생성된 더미 변수들 중 첫 번째 카테고리를 삭제
        data_dummies=pd.get_dummies(data['Survived'], drop_first=True, dtype=int)
        ```

#### 2. 구간화(Binning)
> 연속적인 값을 일정 구간(Bin)으로 구분
<br>

- `pd.cut(데이터프레임['컬럼명'], bins=[나누는 기준 리스트], labels=[지정할 label])`
- 예시 코드
    ```python
    #binning
    count, bin_dividers = np.histogram(df["카테고리"], bins=3) #3개의 구간으로 나눔 
    bin_names = ["상", "중", "하"]
    df["new_bins"] = pd.cut(x=df["카테고리"],
        bins = bin_dividers, # 경계값 리스트 
        labels = bin_names, # bin 이름
        include_lowest = True) # 첫 경계값(여기선 10) 포함 여부 -> True여야 첫번째 데이터 누락 X
    ```
- count, bin_dividers 에 따르는 결과 예시 (이게 pd.cut 할 때 전달되는 것)
    - count = [5, 8, 12] : 각 bin(구간)별 데이터 개수, 자동으로 계산 된 것임
    - bin_dividers = [10, 30, 50, 70] : 3개 구간을 만드는 경계값 : 10~30은 "상", 30~50은 "중"... 이런 식

#### 3. 파생변수 생성
> 기존 피처를 변환해 새로운 피처(변수) 생성
- 목적 : 기존 데이터를 다른 관점에서 분석해 새로운 특성, 정보를 파악
- 원칙이 있지 않아, 분석가의 역량에 영향을 받음
- 방법 : 단위 변환, 표현 형식 변환, 요약 통계량 변환, 정보 추출, 변수 결합, 조건문 이용 등
    - 예시 : `일자` 피처를 변환해 `요일` 피처를 추가

#### 4. 스케일링
> 서로 다른 변수의 값 범위를 일정 수준으로 맞춤
- **정규화** : 숫자의 상대적 크기 차이로 인한 왜곡 방지를 위해, 각 변수의 데이터 값을 동일한 크기 기준으로 나눈 비율(0-1)로 나타내는 작업
- 주의 : 이상치 제거가 선행되어야 함
- **변수 변환** : 단위 변경이 필요하거나 분포가 편향되어 있는 경우, 변수 간 관계가 드러나지 않는 경우 사용
     - 주요 방법 : Log 함수(몰려있는 데이터를 펴줌), 제곱근(변화 폭을 완만하게 조절) 등
     - 사용 라이브러리 : sklearn
        - 제공 스케일러 : StandardScaler, MinMaxScaler, QuantileTransformer, RobustScaler`

1. **<span style="background-color:#fffa7d; color:#000000; padding:0px; border-radius:5px;">StandardScaler (표준화)</span>**<br><br>
    - 각 feature의 **평균을 0, 분산을 1**로 변경 -> 모든 피처가 같은 스케일을 같게 됨
    - **정규분포**를 따른다고 가정하는 기술에 적합
    - 예시 코드
        ```python
        from sklearn.preprocessing import StandardScaler
        scaler = StandardScaler()
        df['스케일링된 새로운 컬럼'] = scaler.fit_transform(df[['기존 컬럼']])
        ```
2. **<span style="background-color:#fffa7d; color:#000000; padding:0px; border-radius:5px;">MinMaxScaler</span>**<br><br>
    - **모든 feature가 0-1 사이에 위치**하게 만듦 -> 다양한 비율의 속성들을 같은 비율로 조정
    - 연산 속도, 알고리즘 최적화에 효과적
    - 예시 코드
        ```python
        from sklearn.preprocessing import MinMaxScaler
        scaler = MinMaxScaler()
        df['스케일링된 새로운 컬럼'] = scaler.fit_transform(df[['기존 컬럼']])
        ```
