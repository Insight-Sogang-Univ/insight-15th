### 1. 판다스란?  
데이터 분석에 최적화된 파이썬 라이브러리  
기본 틀 : DataFrame  
Numpy의 확장격 -> Numpy 기능 + 구조화된 데이터 처리 기능

### 2. 판다스 라이브러리를 불러오는 방법  
 1. 라이브러리 불러오기  
 **import pandas as pd**  
   - **import pandas** - pandas를 실행하겠다  
   - **as pd** - pd라는 이름으로 pandas를 사용하겠다  
2. 데이터 불러오기  
**df=pd.read_csv('파일 경로')** 


### 3. 데이터프레임  
- 컴퓨터로 정보를 정리하고 저장하는 도구  
- 기본 구조 : 행x열 (데이터의 위치를 나타내는 역할)  
- Series
  - 1차원 자료구조  
  - 값(value)와 인덱스(index)로 구성  
  - Series를 구성하는 값들은 모두 같은 데이터타입  
- DataFrame
  - 2차원 배열구조  
  - 여러개의 Series가 모여 하나의 데이터프레임 구성



### 4. 데이터 확인
- 일부 데이터를 미리 확인하여 데이터가 어떤 구조와 값을 가지고 있는지 개괄적으로 파악  
  - df.head()  
  - df.tail()
- 데이터프레임에 몇 개의 행과 열이 있는지, 각 열의 데이터 타입, 결측치의 개수 등을 확인  
  - df.shape()  
  - df.info()  
  - df.describe()  
  - df.isnull()  
  - df.dtypes()
- 데이터의 전체 개수와 고유값 파악  
  - df.count()
  - df[열 이름].value_counts()  
  - df[열 이름].unique()

### 5. 데이터 추출  
- 인덱스와 칼럼을 사용해 추출  
  - indexing / slicing
- 행 단위 추출  
  - loc / iloc
- 열 단위 추출  
  - 열 이름으로 인덱싱
- 불리언 인덱싱
  - T/F를 통해 특정 조건을 만족하는 데이터만 추출


### 6. 데이터 조작  
- 데이터 추가 / 삭제  
  - 추가 : df['새로운 열 이름'] = 리스트  
  - 삭제 : df.drop(labels, axis)  
- 데이터 정렬  
  - df.sort_index() - 인덱스 기준 정렬  
  - df.sort_values(by="열 이름") - 특정 열의 값으로 정렬
- 데이터 변환
  - df[열 이름] = df[열 이름].astype(바꿀 자료형)  
  - df.replace(바꾸고 싶은 값, 바꿀 값)  
  - df.transpose() / df.T()

### 7. 집계함수 사용법
- 집계함수란?
  - 여러 행으로부터 하나의 결과값을 반환하는 함수  
  - 대표적인 집계함수
    - mean(), midian(), std(), sum(), max(), min()  
  - 어느 방향으로 연산을 진행할지 설정
    - 열들에 대한 결과값 -> axis=0 (행방향 연산)  
    - 행들에 대한 결과값 -> axis=1 (열방향 연산)
  - 그룹화
    - gropby 메소드와 함께 사용해 일부 행을 대상으로 집계
