### <span style="color: #000000;background-color: #ffdba4">**1. EDA(Exploratory Data Analysis)란?**</span>  
<U>**데이터 분석(모델링)을 위해 데이터를 여러가지 방식으로 파악하는 모든 과정**</U>  
- 주어진 데이터에 대해 알아보는 방법
    - 시각화 같은 도구를 통해서 패턴 발견
    - 데이터의 특이성 확인
    - 통계와 그래프(혹은 시각적 표현)으로 가설을 검정하는 과정 이용 

##### <span style="color: #000000;background-color: #ffc2e5">   **(1) EDA 대상**</span>  
- 일변량 (Univariate)
    - 한번에 파악하려는 변수 : <U>**1개**</U>  
    - 목적 : 데이터를 설명 & 그 안에 존재하는 패턴 파악 
- 다변량 (Multi-variate)
    - 한번에 파악하려는 변수 : <U>**여러개**</U>  
    - 목적 : 여러 변수들 간의 관계 파악  
    - 개별 데이터를 먼저 파악하는 것이 오류에 대처하기에 용이

##### <span style="color: #000000;background-color: #ffc2e5">   **(2) EDA 종류**</span>  
- 시각화 (Graphic)  
    - 차트 혹은 그림 등을 이용하여 데이터 확인  
    - 대략적 형태 파악 가능
- 비시각화 (Non-Grapic)
    - 주로 Summary Statistics를 통해 데이터 확인  
    - 정확한 값 파악 가능

##### <span style="color: #000000;background-color: #ffc2e5">   **(3) EDA 유형**</span>  
- 일변량 비시각화 (Uni-Non Graphic)
    - 주어진 데이터의 Distribution 확인
- 일변량 시각화 (Uni-Graphic)
    - 주어진 데이터를 전체적으로 파악
- 다변량 비시각화 (Multi-Non Graphic)
    - 주어진 둘 이상의 변수 간 관계 확인
- 다변량 시각화 (Multi-Graphic)
    - 주어진 둘 이상의 변수 간 관계를 전체적으로 확인

### <span style="color: #000000;background-color: #ffdba4">**2. 데이터 확인**</span>


##### <span style="color: #000000;background-color: #ffc2e5">   **(1) 데이터 읽기/쓰기**</span>  
- 절대경로 / 상대경로
    - 파일경로를 나타내는 방법
    - 절대경로 : root파일부터 시작해서 해당파일까지의 모든 경로
    - 상대경로 : 현재 작업하고 있는 디렉터리를 기준으로 상하위 디렉터리와 깉이 상대적으로 표현되는 경로
- 데이터 입출력  

|File Format|Reader|Writer|
|---|---|---|
|**CSV**|read_csv|to_csv|
|JSON|read_json|to_json|
|HTML|read_html|to_html|
|**MS EXCEL**|read_excel|to_excel|
|SQL|read_sql|to_sql|  
- CSV 파일
    - 데이터를 쉼표(,)로 구분하고 있는 텍스트 파일  
    - 데이터의 크기가 작고 압축이 용이 -> 가장 널리 사용
- EXCEL 파일  
    - 행과 열이 데이터프레임의 행, 열로 일대일로 대응


##### <span style="color: #000000;background-color: #ffc2e5">   **(2) 데이터셋 파악하기**</span>  
1. 데이터프레임 확인 : head(n)

2. 데이터 변수(column=feature) 확인
- 변수 정의 확인  
    - 어떤 정보를 가지는 변수인지 확인
- 변수 유형 확인  
    - 질적,범주형 / 양적,수치형
- 변수 데이터 형식(자료형 확인)
    - 날짜, 수치, 텍스트, 이미지  
3. 데이터 분포 확인  
- 단변수 분석
    - 변수들의 분포를 그래프를 통해 하나의 데이터 분포로 확인


### <span style="color: #000000;background-color: #ffdba4">**3. 데이터 전처리**</span>  
<U>**데이터를 분석에 적합한 형태로 만드는 과정**</U> 


##### <span style="color: #000000;background-color: #ffc2e5">   **(1) 결측값 처리**</span>  
<U> 결측값 : 데이터 수집 과정에서 측정되지 않거나 누락된 데이터 </U>
1. 결측값 확인
- 데이터프레임 상에서 Nan=N,?,0으로 나타나 있는 결측값 확인
- pandas function
    - info()/value_counts(dropna=False)/isnull()/notnull()/replace()
2. 결측값 처리
- 삭제 : 데이터가 있는 행 또는 열 삭제
    - dropna() : 결측값이 포함된 레이블 제거
    - 주의 : 결측값이 무작위로 발생한 것이 아닌데 관측치를 삭제하면 왜곡된 모델이 생성될 수 있음
- 대체 : 결측값을 다른 값으로 대체 (평균값, 최빈값 등 대푯값 활용)
    - 일괄 대체 : 모든 변수들 일괄적으로 값은 값으로 대체  
    - 유사 유형 대체 : 유사한 범주에 따른 다른 값으로 대체
    - fillna()

##### <span style="color: #000000;background-color: #ffc2e5">   **(2) 이상치 처리**</span>  
<U>이상치 : 관측된 데이터의 범위에서 많이 벗어난 값</U>  
- 이상치 확인
    - 통계를 통해 확인 : describe()  
    - 시각화를 통해 확인 : BoxPlot()  
    - Z-score를 통해 확인
        - 데이터를 평균(0)과 표준편차(1)로 정규화 -> 평균으로부터 얼마나 떨어져있는지 나타냄
    - Tukey Fences를 통해 확인 : 사분위 범위(IQR)를 기반으로 이상치 판단
        - $Q1-(1.5+IQR)$ 미만  
        - $Q3+(1.5*IQR)$ 초과  
- 이상치 제거 및 처리
    - 전체 삭제 : 이상값이 Human error에 의해서 발생한 경우  
    - 다른 값으로 대체 : 절대적인 관측치의 숫자가 작은 경우
        - 평균 등 다른값으로 대체  
        - 예측모델을 만들고 이상값을 예측한 후 해당 값으로 대체
    - 변수화 : 이상값이 자연 발생한 경우  
    - 리샘플링 : 이상값을 분리해서 모델을 만드는 방법

##### <span style="color: #000000;background-color: #ffc2e5">   **(3) 피처 엔지니어링 (변수 가공)**</span>  
<U>피처(변수)들의 형태를 변형하거나 적절하게 처리하는 과정</U>  
- 레이블인코딩 / 원핫인코딩
    - 레이블인코딩
        - 범주형 변수를 0부터 n-1까지의 숫자로 변환
    - 원핫인코딩
        - 범주형 변수를 이진 벡터로 변환   
        - 표현하고 싶은 단어의 인덱스 : 1, 다른 인덱스 : 0  
        - pd.get_dummies(데이터프레임, columns=[컬럼 리스트])  
- 구간화
    - 연속데이터를 일정한 구간으로 나눠서 분석  
    - pd.cut(데이터프레임['컬럼명'], bins=[나누는 기준 리스트], labels=[지정할 label])
- 파생변수 생성
    - 기존의 피처를 다른 피처로 변환(재정의)하여 변수를 추가
- 스케일링
    - 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업  
    - 정규화 작업 수행 : 각 열(변수)에 속하는 데이터 값을 동일한 크기 기준으로 나눈 비율로 나타냄  
    - 스케일러는 이상치에 민감 -> 이상치 제거가 선행되어야 함  
    - StandardScaler
        - 각 피처의 평균을 0, 분산을 1로 변경  
        - 정규분포를 따른다고 가정하는 기술에 적합
    - MinMaxScaler
        - 모든 피처가 0과 1 사이에 위치하도록 만듦  
        - 연산 속도를 높이고 알고리즘 최적화에 효과적


### <span style="color: #000000;background-color: #ffdba4">**3. 데이터 시각화**</span>  
<U>**데이터의 이해를 돕기 위해 그래픽 요소들을 이용해 데이터를 묘사하고 표현하는 과정**</U>
- 일반적으로 Matplotlib과 Seaborn 사용  
- 시각화 전 데이터의 유형(범주형/수치형) 파악하고 결측값 및 이상치가 있는지 확인해야 함  


##### <span style="color: #000000;background-color: #ffc2e5">   **(1) 파라미터**</span>  
<U>프로그래밍된 함수의 입력값</U>
-  <파마리터 = 인자> 형식으로 전달


##### <span style="color: #000000;background-color: #ffc2e5">   **(2) 다양한 그래프**</span>  
- boxplot(상자 수염 그림)
    - 사분위수와 이상치를 보여주는 그래프
- countplot
    - 범주형 변수의 빈도수를 확인하는 그래프
- histplot
    - 도수분포표의 그래프화  
    - 수치형 변수의 구간별 빈도수를 보여주는 그래프
- displot/kdeplot(커널밀도추정 그래프)
    - 히스토그램을 연속적인 곡선으로 연결한 그래프
- barplot/pointplot
    - 범주형 데이터 값에 따른 수치형 데이터 값의 평균값 제공
- scatterplot(산점도그래프)
    - 두 연속형 변수 간 관계를 시각화하기 위해 사용되는 그래프 
- regplot(회귀선이 추가된 산점도그래프)
    - 두개의 연속 변수 사이의 산점도를 그리고 회귀선을 함께 나타내는 그래프
- catplot
    - 수치형 데이터와 범주형 데이터의 관계를 볼 때 주로 사용
- pieplot
    - 데이터의 부분과 전체 간의 비율을 표현하는 그래프
- heatmap
    - 변수간 상관계수를 직관적으로 볼 수 있는 그래프  
    - 상관계수 : 두 수치형 변수 사이의 상관 관계의 정도를 수치적으로 나타낸 계수
        - 음의 상관계수 : 반비례  
        - 양의 상관계수 : 비례
- violinplot
    - 박스 플롯과 커널밀도추정함수 그래프를 합쳐 놓은 그래프
- pairplot
    - 여러 그래프를 한번에 설정
