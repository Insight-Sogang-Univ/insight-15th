# 판다스 사전 과제

### 판다스
> Pandas<br>

- 데이터 분석에 특화된 파이썬 라이브러리<br>
- DataFrame을 기본틀로 사용<br>
- NumPy 확장판<br><br>

`import pandas as pd` -> Pandas 라이브러리 실행 + 별명 pd 설정<br><br>
`pd.read_확장자()` -> 데이터 불러오기<br>
`pd.to_확장자()` -> 데이터 외부에 저장


### 데이터프레임

> Series<br>

- 값(value)과 인덱스(index)로 이루어진 1차원 자료구조
- 데이터프레임에서 특정 열을 1개만 추출하면 시리즈 객체
- 시리즈를 구성하는 값은 동일한 데이터 타입
    - 숫자와 문자가 섞여 있을경우 문자형(object)으로 처리

**시리즈 만드는 법**
1. `series=[1,2,3,4,5]` -> 리스트 생성
2. `sr=pd.Series(series)` -> pd.Series() 함수로 시리즈 객체 생성
3. `index=['a','b','c','d','e']` -> 인덱스 지정<br><br>


> DataFrame<br>

- 정보를 정리하고 저장하는 2차원 배열 구조
- 행(Index)과 열(Column)으로 구성
- 열 단위로는 동일한 데이터 타입, 행 간 데이터 타입은 달라도 됨

**데이터프레임 만드는 법** 
1. `data={'커피 브랜드':['스타벅스','빽다방','컴포즈'], '아메리카노 가격':[4500,2000,1500]}`-> 딕셔너리 생성
2. `df=pd.DataFrame(data)` -> pd.DataFrame() 함수로 데이터프레임 구성
3. `index=[1,2,3], df=pd.DataFrmae(data,index=index)` -> 인덱스, 컬럼 직접 지정 가능 (기본값:0,1,2...)


### 데이터 확인

> head(),tail()<br>

- `df.head(n)` -> 앞 n개의 행 반환
- `df.tail(n)` -> 뒤 n개의 행 반환

> 데이터 정보 확인<br>

- `df.shape` -> (행 개수, 열개수) 반환
- `df.info()` -> 여러 정보 종합적으로 반환
- `df.describe()` -> 수치형 열들의 통계값 반환
- `df.isnull()` -> 데이터 결측치 여부 반환
- `df.dtypes` -> 열들의 데이터 타입 반환

> 고유값 및 개수 확인<br>

- `df.count()` -> 결측치의 개수 제외한 데이터의 행 수 반환
- `df[열 이름].value_counts()` -> 특정 열에 포함된 값들의 빈도수 세어 반환
- `df[열 이름].unique()` -> 특정 열에 포함된 값들을 종류별로 1개씩만 반환

### 데이터 추출

> 인덱싱&슬라이싱<br>

- *인덱싱(Indexing)*: 단일 항목을 지목 `df[1]`
- *슬라이싱(Slicing)*: 범위를 지정해 여러 항목 추출 `df[start:end]` <br><br>

> 행단위 추출<br>

**loc**<br>
인덱스 이름을 기준으로 행 선택<br>
*끝 인덱스 포함*<br>
- `df.loc[행 이름]` -> 행단위 추출
- `df.loc[행 이름, 열 이름]` -> 값단위 추출
- `df.loc[:,열 이름]` -> 열단위 추출

**iloc**<br>
*끝 인덱스 미포함*<br>
인자로 정수형 인덱스 번호 사용
- `df.iloc[행 순서, 열 순서]`

> 열단위 추출<br>

**열 이름으로 인덱싱**<br>
- `df[열 이름]` -> 시리즈 반환
- `df[[열 이름]]` -> 데이터프레임 자료형 유지
- `df[[열 이름1, 열 이름2..]]` -> 여러 개의 열 데이터프레임으로 반환
- `df[열 이름][행 이름]` -> 특정 열에서 해당 행 반환

> 불리언 인덱싱<br>

**특정 조건 만족하는 데이터 추출**<br>
True, False 배열을 통해 데이터 필터링<br>
- `df[조건문]` -> 조건문 True인 행만 인덱싱
- `df[(조건문1) 논리연산자 (조건문2)]` -> and, or 등 논리연산자로 묶어 사용 가능
- `df.loc[조건문, 열 이름]` -> loc에서도 사용 가능, iloc은 불가

> 인덱스 설정<br>

- `df.set_index(열 이름)` -> 기존 열 중 하나를 인덱스로 설정
- `df.reset_index()` -> 인덱스 리셋하고 기존 인덱스 열에 추가
- `df.reindex(인덱스 배열)` -> 인덱스 재배열, 기존 행도 인덱스 따라 같이 이동

### 데이터 조작

> 데이터 추가, 삭제<br>

- `df['새로운 열 이름']=리스트` -> 특정 리스트를 데이터 프레임에 추가
- `df.drop[labels,axis]` -> 특정 열 삭제

> 데이터 정렬<br>

- `df.sort_index()` -> 인덱스 기준으로 정렬
- `df.sort_values(by='열 이름')` -> 특정 열의 값을 기준으로 정렬

> 데이터 변환<br>

- `df[열 이름]=df[열 이름].astype(바꿀 자료형)` -> 특정 열 자료형 변환
- `df.replace(바꾸고 싶은 값, 바꿀 값)` -> 데이터 값 변환
- `df.transpose()` or `df.T` -> 행열 변환

> 데이터프레임 병합<br>

**concat**<br>
단순히 두 데이터 프레임을 이어붙임<br>
- `df=pd.concat([df1,df2],axis=0)` -> 행방향 결합
- `df=pd.concat([df1,df2],axis=1)` -> 열방향 결합

**merge**<br>
공통 항목을 기준으로 병합<br>
- `df=pd.merge(df1,df2,how='inner',on='~~')` -> 교집합(and) 연산
- `df=pd.merge(df1,df2,how='outer',on='~~')` -> 합집합(or) 연산

### 집계함수
> 집계함수<br>

- 여러 행으로부터 하나의 결괏값 반환
- 결측치는 무시하고 계산

`df['컬럼 명'].mean()`(평균), `df['컬럼 명'].median()`(중앙값),
`df['컬럼 명'].std()`(표준편차), `df['컬럼 명'].sum()`(합계),
`df['컬럼 명'].max()`(최대값), `df['컬럼 명'].min()`(최소값)<br><br>

> 그룹화<br>

**groupby**<br>
- 특정 기준으로 그룹을 묶고 연산 처리
- 집계함수와 함께 사용

- `df.groupby('column명1')['column명2'].집계함수` -> column1로 그룹화, column2에 대해 집계함수 적용하여 시리즈 형태 반환
- `df.groupby('column명1')[['column명2]].집계함수` -> 그룹 별로 column2에 대해 집계함수 적용하여 데이터프레임 형태 반환
