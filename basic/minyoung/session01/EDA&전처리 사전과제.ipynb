# EDA&전처리 사전과제

### 데이터 분석 프로세스

1. 데이터 수집: 다양한 형식의 데이터를 수집
2. 데이터 탐색(EDA): 다양한 각도에서 수집한 데이터의 특징 파악
3. 데이터 전처리: 데이터를 분석에 적합한 형태로 수정
4. 데이터 모델링: 정보를 추출하기 위한 모델 구축

### EDA 개념

> EDA

데이터 분석을 위해 주어진 데이터를 파악하는 모든 과정
- 시각화 -> 패턴 발견
- 특이성 확인
- 통계, 그래프로 가설 검정

1. EDA 대상
- 일변량: EDA를 통해 파악하려는 변수 1개
    - 데이터를 설명하고 패턴을 찾는 것이 주요 목적
- 다변량: 파악하려는 변수 여러 개
    - 여러 변수 간 관계 파악하려는 목적

2. EDA 종류
- 시각화: 차트, 그림 등을 이용하여 데이터 확인
- 비시각화: 그래픽적 요소 없이, 주로 Summary Statistics를 통해 데이터 확인

3. EDA 유형
- 일변량 비시각화: 주어진 데이터의 Distribution 확인
- 일변량 시각화: 데이터 전체적으로 살펴봄
- 다변량 비시각화: 변수 간 관계 확인
- 다변량 시각화: 변수 간 관계를 전체적으로 살펴봄

### 데이터 확인

> 데이터 읽기 쓰기

1. 절대경로, 상대경로
- 절대경로: 처음(root)부터 시작하여 목적지까지 전체적인 경로
- 상대경로: 작업하고 있는 디렉터리를 기준으로 상하위 디렉터리오 같이 상대적으로 표현

2. 데이터 입출력
- 읽기: ```read_(파일형식)```
- 쓰기: ```to_(파일형식)```

3. CSV, Excel
- CSV: 데이터를 쉼표로 구분하는 텍스트 파일
    - 데이터 크기 작음, 압축 용이
- Excel: 행과 열이 데이터프레임의 행, 열로 일대일대응
    -특정 시트 설정 가능 <br><br>

> 데이터셋 파악하기

1. 데이터 프레임 보기
- ```head(n)```: 상위 n개 데이터

2. 데이터 변수 확인
- 변수=데이터 프레임의 Column
- 변수 정의 확인
- 변수 유형 확인: 질적, 범주형 / 양적, 수치형
- 변수 자료형 확인: 날짜, 수치, 텍스트, 이미지 등

3. 데이터 분포 확인
- 산포도, 박스플롯, 히스토리그램 등 활용

### 전처리
데이터 전처리: 데이터를 분석에 적합한 형태로 만드는 과정<br>

> 결측값 처리

결측값: 데이터 수집 과정에서 측정되지 않거나 누락된 데이터<br>

1. 결측값 확인
- 데이터프레임 상에서 ```NaN```, ```?```, ```0``` 등의 값으로 나타나 있는 결측값 확인

2. 결측값 처리
- 삭제: 데이터 있는 행 또는 열 삭제 ```dropna()```
    - 기준: ```axis=0```&```axis=1```<br>
            ```how='any'```&```how='all'```<br>
            ```subset=['column명']```<br>
            ```inplace=False```&```inplace=True```<br>
- 대체: 결측 값을 다른 값으로 대체
    - 일괄 대체: 모든 변수들 일괄적으로 같은 값으로 대체
    - 유사 유형 대체: 범주형 변수들을 활용해 유사한 범주에 따라 다른 값으로 대체
    - ```fillna()```

> 이상치 처리

이상치: 관측된 데이터의 범위에서 많이 벗어난 값<br>

1. 이상치 확인
- 통계를 통해 확인: ```describe()```
- 시각화를 통해 확인: 박스플롯
- Z-score을 통해 확인: 일반적으로 2~3 넘어가는 데이터 이상치
- Tukey Fences을 통해 확인: IQR 기반

2. 이상치 제거 및 처리
- 전체 삭제: 단순 오타, 비현실적 응답 등
- 다른 값으로 대체: 평균, 예측값 등으로 대체
- 변수화: 자연 발생한 경우 그 원인이 되는 변수를 추가해 설명
- 리샘플링: 이상값을 분리해서 모델 생성

> 피처 엔지니어링 (변수 가공)

변수들의 형태를 변형하거나 적절하게 처리<br>

1. 레이블인코딩 vs 원핫인코딩
- 텍스트로 주어진 값을 기계가 이해할 수 있도록 숫자로 변경
- Label Encoding: 범주형 변수를 0~n-1까지 숫자로 변환
- One-Hot Encoding: 범주형 변수를 이진 벡터로 변환

2. 구간화
- 일정한 구간으로 나눠서 분석

3. 파생변수 생성
- 기존의 피처를 다른 피처로 변환(재정의)하여 변수 추가
- 분석가의 도메인 지식과 데이터 특성에 대한 이해도 중요

4. 스케일링
- 변수의 값 범위를 일정한 수준으로 맞춤
- 표준화: 평균을0, 분산을 1로 변경하여 모든 피처 같은 스케일
- MinMaxScaler: 모든 피처가 0과 1사이에 위치하게 만듬 (비율 맞춤)

###  데이터 시각화
데이터를 **그래프, 차트, 그림, 도표** 등 시각적 요소로 변환하여 **패턴, 관계, 추세** 등을 파악<br>
- Matplotlib, Seborn 라이브러리 활용

> 파라미터

- 매개변수(parameter): 프로그래밍된 함수의 입력값
- ```파라미터=인자``` 형식으로 전달

> 다양한 그래프

1. boxplot
- 사분위수, 이상치
 <br>

2. countplot, histplot
- countplot: 범주형 변수의 빈도수 확인
<br>
- histplot: 도수분포표, 수치형 변수의 구간별 빈도수
 <br>

3. displot, kdeplot
- 히스토그램을 연속적으로 곡선으로 연결
 <br>

4. barplot, pointplot
- 범주형 데이터(x값)에 따른 수치형 데이터(y값)의 평균 값 제공
 <br>

5. scatterplot, regplot
- scatterplot: 두 연속형 변수 간 관계를 시각화, 데이터 포인트를 분산하여 그림
 <br>
- regplot: 두 개의 연속 변수 사이의 산점도를 그리고 회귀선을 함께 나타냄
 <br>

6. catplot
- 수치형 데이터와 범주형 데이터의 관계
 <br>

7. pieplot
- 데이터의 부분과 전체 간 비율을 표현
 <br>

8. heatmap
- 변수간 상관계수
 <br>

9. violinplot
- 박스 플롯과 커널밀도추정 함수를 합침
 <br>

10. paiplot
- 여러 그래프 한 번에 설정
