# 1. Pandas란?

* 데이터 분석에 특화된 파이썬 라이브러리
* 행과 열로 이루어진 데이터 객체를 만들 수 있으며, 보다 안정적으로 대용량의 데이터를 편리하게 처리할 수 있다.
* numpy의 확장판이다.

## 2. Pandas 라이브러리 호출 방법

* import pandas 로 라이브러리 호출 후 pd.DataFrame()으로 데이터 프레임을 만든다.

## 3.데이터 프레임이란?

* 엑셀처럼 행과 열로 이루어진 2차원 표(Table)형식의 데이터를 다루는 자료구조이다.
* 1차원 데이터인 시리즈가 모여 구성되며 데이터 분석 및 전처리에 필수적으로 사용된다.
* 구조:
    - index:각 행을 식별하는 고유 라벨
    - columns:데이터의 열 이름
    - data: DataFrame 내의 실제 데이터 값
    - dtype: 각 열의 데이터 타입

## 4. 데이터 확인 방법

* info, describe, head로 데이터 파악
* shape로 데이터 크기 확인
* value_counts로 각 열의 값에 대한 모든 발생 횟수를 반환

## 5. 데이터 추출 방법

* 추출 단위: 단일/다중 열단위/행단위 추출, 행과 열 복합 조겆 추출
* 추출 방법: 인덱싱, 슬라이싱, loc, iloc, 불리언 인덱싱 
    - loc을 이용하여 추출 하는 경우에는 이름 제공
    - iloc을 이용하여 추출 하는 경우 숫자로 범위 제공
    - 단일 조건 불리언 인덱싱: df[df['Score']]>=50
    - 다중 조건 불리언 인덱싱: 논리 연산자(&,|,~) 사용

## 6. 데이터 조작 방법

* 데이터프레임 연산
* 데이터프레임 삭제-drop 사용
* 데이터프레임 병합-concrat, merge 사용 
    - concrat: 단순 이어 붙이기. 이때 axis 주의
    - merge: 공통 분모로 합치기

## 7. 집계함수

* 데이터베이스 환경에서 여러 행(row)의 데이터를 입력받아 합계, 평균, 개수 등 요약된 단일 결과값을 반환하는 함수
* count(),min(),max(),sum(),mean() 등을 활용.
* groupBy()와 함께 사용하여 특정 카테고리별의 통계적 데이터 계산.
