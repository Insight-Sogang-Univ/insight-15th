# 2. EDA 개념

- 데이터 분석(모델링)을 위해 데이터를 여러 가지 방식으로 파악하는 모든 과정
- 히스토그램, 산점도와 같은 시각화 도구를 통해 데이터 패턴 및 특이성을 파악하고,통계와 그래프(혹은 시각적 표현)으로 가설을 검정하는 과정 등을 통해 주어진 데이터에 대해 이해함.
- EDA 유형: 일변량 비시각화, 일변량 시각화, 다변량 비시각화, 다변량 시각화

#### 3.2.1 데이터의 종류
- 크게 범주형, 수치형으로 나뉨
- 범주형: 몇 개의 범주 또는 항목의 형태로 나누어 진 자료
    - 명목형: 성별, 혈액형과 같이 **아무런 순서가 없고 단순히 분류를 목적**으로 하는 데이터

    - 순서형: 난이도와 평점 같이 **여러가지 범주에 순서가 주어지는** 자료 

- 수치형:  숫자 형태로 측정되는 자료. 예) 키, 몸무게, 판매량, 불량품 수
    - 이산형 자료: 판매량, 불량품 수 처럼 **데이터와 데이터 간 끊어짐이 있고** 이산적인 값을 가지는 데이터
    
    - 연속형 자료: 키, 몸무게, 나이처럼 값이 **데이터와 데이터 사이에 끊어짐이 없어** 연속적인 값을 가지는 데이터

# 4. 결측값

- 결측값: 데이터 수집 과정에서 측정되지 않거나 누락된 데이터
- 결측값 처리 과정: 
    - 1) 결측값 확인-데이터프레임 상에서 NaN, ?, 0 등의 값으로 나타나 있는 결측값을 확인. 
        - info(), value_counts(dropna=False), isnull(), notnull(), replace() 함수 등을 활용
    - 2) 결측값 처리-삭제, 대체
        - 삭제 매소드-dropna() 
        - 대체 매소드-fillna()

# 5. 이상치

- 이상치: 관측된 데이터의 범위에서 많이 벗어난 값.
- 이상치 처리 과정: 
    - 1) 이상치 확인.
        - 방법:  a. 통계 b. 시각화 c. Z-score d. Tukey Fences
    - 2) 이상치 제거 및 처리
        - 방법: a. 전체 삭제 b. 다른값으로 대체 c. 변수화 d.리샘플링

# 6. 피처 엔지니어링(변수 가공)

- 정의: 해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 피처(변수)들의 형태를 변형하거나 적절하게 처리하는 과정
- 방식: 
    1) 레이블인코딩, 원핫 코딩: 텍스트로 주어지는 값들을 기계가 이해할 수 있는 숫자로 바꾸는 작업
        - Label Encoding: 범주형 변수를 0부터 N-1까지의 숫자로 변환 ex) A,B,C->0,1,2
        - One-Hot Encoding: 범주형 변수를 이진 벡터(0과 1)로 변환. 
        이때 표현하고 싶은 단어의 인덱스는 1, 나머지는 0.  ex) A,B,C->1,0,0/ 0,1,0/ 0,0,1

    2) 구간화: 연속 데이터를 일정한 구간으로 나눠서 분석하는 것.

    3) 파생변수 생성: 기존 데이터에서 새로운 특성과 정보를 얻기 위해 기존의 피처를 다른 피처로 변환하여 변수를 추가하는 과정

    4) 스케일링: 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업. 
    숫자의 상대적인 크기 차이로 인한 결과의 왜곡을 방지하기 위해, 각 열(변수)에 속하는 데이터 값을 **동일한 크기 기준으로 나눈 비율로 나타내는 정규화 작업** 수행
        - 정규화 방식: a. StandardScaler b. MinMaxScaler
