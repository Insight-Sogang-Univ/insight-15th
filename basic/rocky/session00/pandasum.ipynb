# Pandas 개념 및 라이브러리 호출 #
### 개념
- Pandas는 구조화된 데이터를 빠르고 효율적으로 처리하기 위해 설계된 파이썬 핵심 데이터 분석 라이브러리
- DataFrame을 기본 구조로 사용
- Numpy의 확장판

### 호출방법
- import pandas as pd

# 데이터프레임 (DataFrame)
- 행(Row, Index)과 열(Column)로 구성된 2차원 테이블 형태의 자료구조
- 1차원 배열(Array)을 나타내는 Series 객체들이 인덱스(Index)를 기준으로 결합된 것

# 데이터 확인/추출/조작
### 데이터 확인
- df.head(5): 데이터의 상위 5개 행을 출력하여 구조 파악
- df.info(): 결측치(Null) 여부 및 컬럼별 데이터 타입(Dtype) 확인
- df.shape: 데이터프레임의 차원 크기 (행, 열) 튜플 반환
- df.describe(): 수치형 변수의 요약 통계량(평균, 표준편차, 4분위수 등) 산출
### 데이터 추출 
- 인덱싱
- 슬라이싱
- loc와 iloc
- 불리언 인덱싱
### 데이터 조작
- 추가 :  df ['새로운 열 이름'] = 리스트
- 삭제 : df.drop(labels, axis)
- 정렬 : df.sort_index()
- 변환 : df. replace (바꾸고 싶은 값, 바꿀 값)
- 데이터프레임 병합 : 행 방향 - df = pd.concat([df1,df2], axis=0) 
- 데이터프레임 병합 : 열 방향 - df = pd.concat([df1,df2], axis=1)

# 집계함수
- 여러 행으로부터 하나의 결과값을 반환하는 함수 
- 대표적으로 평균,  중앙값, 표준편차, 합계, 최대/최소 값이 있음
### 그룹화
- groupby 메소드는 특정 기준으로 그룹을 묶고, 연산처리를 할 수 있게 해줌
- df.groupby('column명1')[’column명2'].집계함수
- column1로 그룹화를 하고, 그룹 별로 column2에 대해 집계함수를 적용해 결과를 시리즈 형태로 반환
