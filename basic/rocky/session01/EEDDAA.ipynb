# EDA & 전처리
## EDA 개념
- 데이터 탐색(EDA): 다양한 각도에서 수집한 데이터의 특징 파악 (ex. 도표, 그래프, 통계 등)
- 데이터의 패턴이나 특이성 확인
- 향후 분석 과정으로 나아감에 있어 중요한 단계
## 데이터셋 파악하기
1. 데이터 프레임 보기
: head(n): 데이터 프레임 상위 n개 데이터 보여줌
2. 데이터 변수 확인
:  변수 = 데이터 프레임의 column = feature
3. 데이터 분포 확인
: 여러 raw data 변수들의 분포를 단변수(그래프 등)로 확인할 수 있음 - 데이터 전처리 과정이 필요함
## 결측값 & 이상치 처리
### 결측값 처리
- 확인: Nan, ?, 0 등의 값으로 나타나 있는 결측값을 확인
- 삭제: 데이터가 있는 행 또는 열을 삭제 - dropna()
- 대체: 결측값을 다른 값으로 대체 - fillna()
### 이상치 처리
- 이상치: 데이터의 범위에서 많이 벗어난 값
- 확인: describe(), BoxPlot, Z-core, Tukey Fences
- 제거 및 처리: 전체 삭제, 다른 값으로 대체, 변수화, 리샘플링
## 피처 엔지니어링 (변수 가공)
- 해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 피처(변수)들의 형태를 변형하고 처리
1. 레이블인코딩, 원핫인코딩: 문자열 값들을 숫자형으로 변경하는 작업
2. 구간화(Binning): 연속적인 값을 일정한 구간(Bin)으로 구분 
3. 파생변수 생성: 기존의 피처를 변환해 새로운 피처 생성
4. 스케일링: 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업

## 데이터 시각화
