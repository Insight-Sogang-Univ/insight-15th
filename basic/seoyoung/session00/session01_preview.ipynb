# 데이터 추출

- 열 단위 추출
    - 단일 칼럼: 인덱싱 활용(->값이 시리즈 형태로 반환됨), loc 활용, iloc 활용
    - 다중 칼럼: 인덱싱 활용, loc 활용, iloc 활용
                *여러 개의 칼럼을 동시에 인덱싱할 때는 무조건 리스트 형태로 제공*
- 행 단위 추출
    - 슬라이싱(범위 지정)
    - loc: 인덱싱과 슬라이싱 모두 가능. **끝 인덱스 포함**
    - iloc: **끝 인덱스 미포함**
- 행과 열 복합 조건 추출
    - 같은 방법으로 하되, loc[(행 범위), (열 범위)] 형식


- 불리언(boolean) 인덱싱: 값이 Trued인 것만 추출, 값이 1인 것만 추출 등
                        - 논리연산자 사용: (조건문1) &/| (조건문2)

# 인덱스 설정

- set_index(): 특정 열을 인덱스로 사용. 인덱스 지정 안 하면 0,1,2,... 처럼 숫자 인덱스 지정됨
- sort_index(): 인덱스를 기준으로 오름차순 정렬
- reset_index(): 인덱스 리셋. 인덱스 지정 안 한 상태로 복구됨

# 데이터 조작

- 데이터 추가: df[(새 column)] = (값)
- 연산: df[열 이름] +/-/...
- 데이터 삭제: df.drop(열 이름, axis)
- 데이터 정렬: sort_index()에서 ascending = True/False 설정
- 데이터 병합: pd.concat([d1, d2], axis)

# 집계함수

- count(): axis=0이 기본 -> 해당 col의 행 개수 세기
- mean(): axis=0이 기본 -> 해당 col의 평균값
- 이 외에도 max(), min(), sum() 등이 있다
- groupby(): 그룹화된 객체 생성. but 여기서 집계함수를 적용해야 인사이트 도출 가능
