## **EDA**(Exploratory Data Analysis) 란?

> 개념  
- 데이터 분석(모델링)을 위해 **데이터를 여러 가지 방식으로 파악하는 모든 과정**  
    <방법>
    - 시각화를 통한 패턴 발견
    - 데이터 특이성 확인
    - 통계/그래프 등으로 가설 검정
> 대상 
- 일변량 EDA: 데이터 설명 및 패턴 파악  *Univariate라고도 함*
- 다변량 EDA: 여러 변수 간 관계 파악    *Multi-variate라고도 함*
> 종류  
- 시각화: 차트나 그림 이용->한눈에 파악 가능
- 비시각화: 통계값 이용->정확한 값이 필요할 때
> 유형  
- 일변량 비시각화: 데이터의 distribution 확인
- 일변량 시각화: 데이터를 전체적으로 파악
- 다변량 비시각화: 변수 간 관계 확인
- 다변량 시각화: 변수 간 관계를 전체적으로 파악


## 데이터 확인
> 데이터 읽기/쓰기  
- 읽기 = 불러오기, 쓰기 = 만들기
- 절대 경로: root~목적지    *C:\ 부터 시작*

    - 상위: ../
    - root: /
- 데이터 입출력  
  
*csv는 쉼표로 구분된 텍스트 파일, excel은 행과 열로 표현된 파일*  
> 데이터셋 파악하기  
1. 데이터 프레임 보기
2. 데이터 변수(feature) 확인: 정의, 유형, 자료형 확인
    - 범주형: value가 문자로 표현. 순서 유무에 따라 순서형/명목형
    - 수치형: value가 숫자로 표현. 이산형/연속형
3. 데이터 분포 확인하기: 시각화를 통해 가능

## 전처리
> 개념  
- 데이터를 분석에 적합한 형태로 만드는 과정(이상값, 결측값, 노이즈 수정)
> **결측값 처리**  
- 결측값: 누락 데이터
- 방법 1 ) 삭제
    - 기준: axis = 0 또는 axis = 1
    - 조건: how = 'any'(하나라도 결측값이 있으면 삭제) 또는 'all'(모두 결측값일 때만 삭제)
- 방법 2 )  대체 (평균값, 최빈값 등의 대푯값으로 대체)
---  
> **이상치 처리**  
- 이상치: 데이터 범위에서 많이 벗어난 값
- 통계, 시각화, Z-score, Turkey Fences을 통해 확인
    - Z-score: 데이터를 평균(0)과 표준편차(1)로 정규화하여 평균으로부터 얼마나 떨어져있는지 표현
    - Turkey Fences: 사분위 범위를 기준으로 이상치 판단(상위 25%, 하위 25%)
- 방법 1) 전체 삭제
- 방법 2) 대체
    - 다른 값(평균 등)으로 대체
    - 결측값과 유사하게 다른 변수들을 사용해 예측 모델 생성, 이를 통해 이상값 예측 후 대체
- 방법 3) 변수화
    - 새로운 변수를 생성함으로써 이상값 유지
- 방법 4) 리샘플링
    - 이상값을 분리해서 모델 생성
---
> **피쳐 엔지니어링(변수 가공)**  
- 해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 변수들의 형태를 변경하거나 적절하게 처리(기존 데이터에 정보 추가)
- 방법 1) 레이블인코딩, 원핫인코딩
    - 레이블인코딩: 범주형 변수를 0~N-1까지의 숫자로 변환
    - 원핫인코딩: 범주형 변수를 이진 벡터(0/1)로 변환
    => 컴퓨터가 텍스트를 이해할 수 있게 숫자로 바꾸는 것  
- 방법 2) 구간화
    - 연속 데이터를 일정 구간으로 나눠서 분석
- 방법 3) 파생변수 생성
    - 기존의 피처를 다른 관점에서 분석해 새로운 피처로 변환(재정의)
- 방법 4) 스케일링
    - 정규화 작업(StandardScaler, MinMaxScaler)을 통해 숫자의 상대적 크기로 인한 왜곡 방지

## 데이터 시각화
- 정보를 그래프, 차트, 그림 등으로 변환 -> 패턴, 관계, 추세 한눈에 파악
- Matplotlib, **Seaborn** 사용
- 그래프의 종류
    - boxplot: 사분위수, 이상치 표현  
    *showfilers=False로 설정하면 이상치 표시 안 함*
    - countplot: 범주형 변수의 빈도수 확인(막대그래프)
    - histplot: 수치형 변수의 구간별 빈도수 확인(막대그래프)
    - displot, kdeplot: 커널밀도추정 그래프. 히스토그램을 연속적인 곡선으로 연결한 것
    - barplot: 범주형 데이터 값 x에 따른 수치형 데이터 값 y의 **평균값**
    - pointplot: 막대그래프와 동일한 정보 제공
    - scatterplot: 두 연속형 변수 간의 관계 표현(산점도 그래프)
        - regplot: 산점도에 회귀선을 추가한 그래프
    - catplot: 수치형 데이터와 범주형 데이터의 관계 파악
    - pieplot: 데이터의 부분과 전체 간의 비율 표현(원그래프)
    - heatmap: 변수간 상관계수 표현
    - violinplot: 박스플롯 + 커널밀도추정함수
    - pairplot: 여러 변수 간의 산점도를 한 번에 보여주는 그래프
