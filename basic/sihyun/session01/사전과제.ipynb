# 판다스 라이브러리 불러오는 방법




 import pandas as pd
 df=pd.read_csv('파일 경로')



# 데이터 프레임
컴퓨터로 정보를 정리하고 저장하는 도구 (series는 1차원, DataFrame은 2차원)

행과 열로 구성되어 있음 -> 데이터 타입이 달라도 됨

(참고)시리즈는 1차원으로서 array임 -> 시리즈를 구성하는 값들은 모두 같은 데이터 타입이어야 함





# 데이터 확인, 추출, 조작하는 방법

## 확인
(1) 앞 뒤 확인
df.head(n): 앞 n개 행 반환
df.tail(n): 뒤 n개 행 반환

(2) 정보 확인
df.shape: 행, 열 개수 반환
df.info(): 여러 정보 반환
등등

(3) 고유값 및 개수 확인
df.count(): 결측치의 개수를 제외한 데이터의 행 수 반환
등등


## 추출
(1) 인덱싱: 단일 항목을 지목할 때 사용
(2) 슬라이싱: 범위를 지정해 여러 항목을 추출할 때 사용
(3) loc: 행을 선택할 때
(4) iloc: 정수형 위치를 사용할 때
(5) 열단위 추출: 컬럼명
(6) 불리언 인덱싱: T/F의 배열을 통해 데이터를 필터링하는 연산
(7) 인덱스 설정" 기본적으로 0부터 시작하는 정수형 인덱스를 가지지만 원하는대로 변경이 가능함

## 조작
(1) 추가: 특정 데이터 셋을 열로 추가 -> df['새로운 열 이름']=리스트
(2) 삭제: df.drop(lables, axis)
(3) 정렬: df.sort_index(): 인덱스 기준으로 정렬 & df.sort_values(by="열 이름"): 특정 열의 값을 기준으로 정렬
(4) 변환: df.dtypes로 데이터 타입 변경, df.replace로 값 변환, df.transposw로 값 전치
(5) 병합: concat -> 단순 병합, merge -> 공통 항목을 기준으로 변합

# 집계함수

여러 행으로부터 하나의 결괏값을 반환하는 함수
-> 각 컬럼의 평균 등 대표값을 구할 때 주로 사용 (mean, medium, std 등)

groupy 메소드는 특정 기준으로 그룹을 묶고 연산 처리를 하게 함 -> 보통 집계 함수와 함께 사용
