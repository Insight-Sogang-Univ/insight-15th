# EDA 개념

다양한 각도에서 수집한 데이터의 특징을 파악하는것 (특성,패턴,관계 등)을 파악하고 시각화하면서 데이터를 이해하고 전처리, 모델링 등의 후속 분석에 대한 방향성 제시  

시각화 같은 도구를 통해 패턴을 발견하거나 데이터의 특이성을 확인하거나 통계와 그래프로 가설을 검증하는 과정을 통해 주어진 데이터에 대해 알아 보는것

# 데이터셋 파악

1. 데이터 프레임
head(n)

2. 데이터 변수 확인
어떤 정보를 가지고 질적-범주형과 양적-수치형으로 구분, 데이터 형식을 확인

3. 데이터 분포 확인
단변수 분석은 원시 데이터의 평균, 최빈, 중앙값등 변수들의 분포를 다양한 그래프를 통해 단변수를 확인할 수 있는 방법임

# 전처리
데이터를 분석에 적합한 형태로 만드는 과정

# 결측값 처리
1. 확인
NaN, ?, 0 등의 값으로 나타나있는 결측값을 확인한다
info, insull 등 다양한 함수로 이를 확인 가능

2. 처리
2.1 제거
dropna 를 통해 결측값이 포함된 레이블을 제거 단. 결측값이 무작위로 발생한 경우에 사용해야 하며 그렇지 않을 경우 왜곡된 모델이 생성될 수 있음

2.2 대체
결측값을 다른 값으로 대체(평균, 최빈값등)
fillna 함수를 사용해 일괄 or 유사 유형 대체를 실시

# 이상치 처리
1. 확인
통계를 통해 확인
2. 시각화를 통해 확인
3. z-score를 통해 확인
4. Tuckey Fences를 통해 확인
5. 제거 및 처리
5.1 전체 삭제
휴먼에러에 의해서 발생한 경우에 해당 관측치를 삭제
5.2 대체
절대적인 관측치의 숫자가 작은 경우 단순 삭제를 하면 신뢰성 문제가 생기기겡 다른 값으로 대체
5.3 변수화
이상값이 자연 발생한 경우 이상값에 대해 파악해 변경
5.4 리샘플링
이상값을 분리해서 모델을 만드는 방법이 존재

# 피처 엔지니어링
기존의 데이터에 정보를 추가하는 일련의 과정을 변수 가공(피처 엔지니어링)이라고 함
이는 기존의 데이터를 보다 유용하게 만듦

1. 레이블 코딩: 범주형 변수를 0부터 N-1까지의 숫자로 변환 vs 원핫 코딩: 범주형 변수를 이진수로 변경
2. 구간화: 연속 데이터를 일정한 구간으로 나누어서 분석
3. 파생 변수 생성: 기존의 피처를 다른 피처로 변환하여 변수를 추가함
4. 스케일링: 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업으로 정규화 작업을 수행함

# 데이터 시각화
숫자와 텍스트를 그래프, 차트, 그림, 도표 드으이 시각적 요소로 변환하여 데이터 패턴, 관계, 추세를 파악

Metplitib와 Seaborn을 주로 이용

boxplit, countplot, histplot, displot, kdeplot, barplot, pointplot, scatterplot, regplot, catplot, pieplot 등으 존재함
