# 판다스 사전 과제

## 판다스의 개념과 판다스 라이브러리를 불러오는 방법  
- **판다스란?** 데이터 **분석**에 특화된 파이썬 라이브러리  
  - 행과 열로 이루어진 **테이블 형태**의 데이터 구조인 *DataFrame*을 기본 틀로 사용   
  
    
    
- **판다스 라이브러리 불러오기**  
1. import pandas as pd 활용  
2. 데이터 불러오기: pd.read_csv()


## 데이터프레임이 무엇인지
- **데이터 프레임이란?** 컴퓨터로 정리하고 저장하는 도구/ 판다스에서 사용하는 주요 데이터 구조/ $=!$시리즈  
  - **가로** 방향의 '행'과 **세로** 방향의 '열'로 구성  
  - 행렬은 데이터프레임 내에서 **데이터의 위치를 나타내는 역할** 수행  
  - **2차원** 배열 구조
  - 컬럼 간의 데이터 타입은 달라도 무방
  

- **Series란?** '값'과 '인덱스'로 이루어진 가장 간단한 **1차원** 자료구조
  - 인덱스가 있는 1차원 배열
  - 구성하는 값들이 모두 *같은* 데이터 타입이어야함

## 데이터를 확인, 추출, 조작하는 방법을 간단하게 작성  
- **데이터 확인**  
  **1. head(),tail()** : 앞/뒤 일부 데이터를 미리 확인하여 데이터의 **개괄적인** 타입 파악  
  **2. 데이터 정보 확인**  
    - df.shape(): (행 개수, 열 개수) 반환  
    - df.info(): 여러 정보들의 종합 반환  
    - df.describe(): 수치영 열들의 기본적인 통계값 반환  
    - df.isnull(): 데이터의 결측치 여부 반환  
    - df.dtypes: 열들의 데이터타입 반환   
  **3. 고유값 및 개수 확인**
    - df.count(): 결측치의 개수 제외, 데이터의 행 수 반환  
    - df[열 이름].value_counts(): 특정 열에 포함된 **값**들의 *빈도수* 반환  
    - df[열 이름].unique(): 특정 열에 포함된 **값**들을 *종류별로 1개씩만* 반환

- **데이터 추출**  
  **1. 인덱싱&슬라이싱**: 인덱스와 칼럼을 사용해 데이터를 추출하는 것  
     - 인덱싱: 단일 항목 **지목**에 사용  
     - 슬라이싱: **범위**를 지정해 여러 항목의 추출에 사용  
**2. 행단위 추출: loc과 iloc**  
     - **인덱스 이름** 기준 선택: loc  
     -**정수형 위치** 기준 선택: iloc  
     - 차이점: loc은 끝 인덱스를 포함하여 반환하는 반면, iloc은 끝 인덱스 미포함  
**3. 열단위 추출: 컬럼명**  
     - 열 이름으로 인덱싱  
**4. 불리언 인덱싱**: *특정 조건에 만족하는 데이터*를 추출하고 싶을 때 사용  
     - True 또는 False의 배열을 통해 데이터 필터링  
     - 조건문 여러 개에 대한 논리 연산자 사용 가능  
     - loc에서 사용 가능  
**5. 인덱스 설정**  
     - 기본값: 0부터 시작하는 정수형 인덱스   
     - 변경 가능

- **데이터 조작**  
**1. 데이터 추가, 삭제**  
  - 특정 데이터 셋을 열로 추가 및 삭제 가능  
**2. 데이터 정렬**: 새로운 방식으로 가능  
**3. 데이터 변환**  
  - 특정 컬럼 내 데이터의 *자료형* 변환 가능  
  - 데이터의 *값 자체* 변환 가능  
  - **전치**를 통해 행과 열 교환 가능  
**4. 데이터 프레임 병합**  
   1) concat: 단순 병합/보통 동일 구조 병합에 사용  
   2) merge: *공통 항목* 기준 병합/ 형태 달라도 겹치는 데이터 기준 병합 가능  
  

## 집계함수를 사용하는 방법  
- **집계함수**: 여러 행으로부터 하나의 결괏값을 반환하는 함수  
  - 연산을 통한 대표값 도출에 주로 사용  

- **그룹화**
  - groupby: 특정 기준으로 병합, 연산 처리 가능하게 함
