# 2차 교육세션 사전과제 - EDA & 전처리

## EDA 개념


* EDA=데이터 탐색 
  * 데이터를 여러 가지 방식으로 파악하는 모든 과정 
  * 다양한 각도에서(도표, 그래프, 통계) 수집한 데이터의 특징 파악 
  * 시각화 하며 데이터 이해 -> 후속 분석에 대한 방향성 제시  

### 데이터 전처리  
: 데이터를 분석(모델링)에 적합한 형태로 손질하는 과정  
* 이상치 제거, 결측치 처리, 노이드 제거 등의 작업 수행  
* 데이터의 왜곡 피하고, 데이터 신뢰도 높이는 과정  
* 반복 수행  

### 주요 개념  
(1)EDA 대상(일변량/다변량)
> 다변량의 경우 변수의 동시 확인 전, 개별 데이터 먼저 파악  
   -> 오류 대처에 용이  

(2) EDA 종류(시각화/비시각화)
> 시각화: 시각적 파악  
  비시각화: 수치적 파악/정확한 값이 필요한 경우  

(3) EDA 유형  
  * 일변량 비시각화  
  * 일변량 시각화  
  * 다변량 비시각화  
  * 다변량 시각화 


## 데이터셋 파악  
(1) 데이터 프레임 보기  
(2) 데이터 변수 확인 : 정의, 유형, 자료형 확인  
  * 데이터의 종류: 범주형(명목형, 순서형), 수치형(이산형, 연속형)  
(3) 데이터 분포 확인


## 결측치 & 이상치 처리  
* **결측치**: 데이터 수집 과정에서 측정되지 않거나 누락된 데이터  
  -> 존재할 시, 변수 간 관계의 왜곡 발생하여 모델의 정확성 하락 가능 -> 데이터의 시각적 표현에도 문제  

  (1) 결측치 확인: NaN, ?, 0 등의 값으로 표현  
  (2) 결측치 처리: 삭제, 대체  


* **이상치**: 관측된 데이터의 범위에서 많이 벗어난 값  
-> 미처리 시, 데이터 분석에 큰 영향  
  (1) 이상치 확인: 통계, 시각화, Z-score, Turkey Fences 통해  
  (2) 이상치 제거 및 처리: 전체 삭제, 다른 값으로 대체, 변수화, 리샘플링



## 피처 엔지니어링(변수 가공)  
: 피처들의 형태 변형 혹은 적절하게 처리하는 과정  
* **변수 가공**: 기존의 데이터에 정보를 추가하는 일련의 과정  
  - 데이터 전처리의 마지막 단계  
  - 기존의 데이터를 보다 유용하게 만드는 방법  
  - 도메인 지식 여부에 따라 변수 가공의 결과와 유의성의 변동성 큼  

* **피처 엔지니어링 방식**  
 1) 레이블인코딩, 원핫인코딩  
 2) 구간화  
 3) 파생변수 생성  
 4) 스케일링

## 데이터 시각화  
: 그림, 도형 등 그래픽 요소들을 이용해 데이터를 묘사하고 표현하는 과정  
-> 패턴, 관계, 추세 등 쉽게 파악 가능

 - 결측값과 이상치 확ㅇ니  
 - 그래프의 유형 선택
