# 판다스

판다스란 panel datas의 약어로 데이터 분석에 특화된 파이썬 **라이브러리**

라이브러리란 **미리 만들어진 기능**들이 들어 있는 도구 상자  >import pandas as pd

### 데이터 프레임

컴퓨터로 정보를 정리하고 저장하는 도구

데이터프레임의 기본 구조는 ‘행(Index)’과 ‘열(Column)’로 이루어짐

함수의 (x,y) 좌표처럼, 행과 열 또한 데이터프레임 내에서 데이터의 위치를 나타내는 역할

### Series

시리즈는 ‘값(value)’과 ‘인덱스(index)’로 구성된, 가장 간단한 1차원 자료구조

cf. 데이터프레임은 행과 열로 이루어진 2차원 배열 구조

### 데이터 확인, 추출, 조작법

1. 데이터 확인

head(), tail() - 앞/뒤 일부 데이터를 미리 확인하여 데이터가 어떤 구조와 값을 가지고 있는지 개괄적으로 파악

 몇 개의 행과 열이 있는지, 각 열의 데이터 타입, 결측치의 개수 등을 확인

데이터의 전체 개수와 고유값

2. 데이터 추출

분석에 필요한 부분만 골라내어 가공

- 인덱싱(Indexing) & 슬라이싱(Slicing)

인덱싱은 단일 항목을 지목할 때 사용, 슬라이싱은 범위를 지정해 여러 항목을 추출할 때 사용

- 행단위 추출 : loc과 iloc

인덱스 이름(index label)을 기준으로 행을 선택할 때는 `loc`을 이용하고, 정수형 위치(integer position)를 사용할 때는 `iloc`을 이용

3. 데이터 조작

- 데이터 셋을 추가하거나 삭제 가능

- 데이터를 새로운 방식으로 정렬 가능

- 데이터의 자료형 변형 가능

- 데이터 프레임을 병합 가능(merge)

## 집계함수

- 여러 행으로부터 하나의 결괏값을 반환하는 함수

- `df['컬럼 명'].mean()` : 평균
- `df['컬럼 명'].median()` : 중앙값
- `df['컬럼 명'].std()` : 표준편차
- `df['컬럼 명'].sum()` : 합계
- `df['컬럼 명'].max()` : 최대값
- `df['컬럼 명'].min()` : 최소값
