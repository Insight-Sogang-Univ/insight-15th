0. 데이터 분석 프로세스



- **데이터 수집** - 다양한 형식의 데이터(정형/ 비정형/ 반정형 데이터)
- **데이터 탐색** (EDA-exploratory data analysis) - 데이터의 핵심 특성, 패턴 등을 파악 (비유: 요리 전 재료 파악)

- **데이터 모델링** - 유용한 정보를 추출하기 위해 모델을 구축! - 예측, 분류, 군집화 등의 목적에 따라 모델을 선택하고 학습시키는 과정을 의미 (찐 요리 단계!)



## 1. **EDA**의 개념
- 데이터 분석(*모델링*)을 위해 데이터를 여러 가지 방식으로 파악하는 모든 과정

a. *시각화*같은 도구를 통해서 *패턴*을 발견하거나

b. 데이터의 **특이성**을 확인하거나 (특이성: *이상치*, *결측치* ) ~~치치거리니까 다마고치 이름같음~~

c. 통계와 그래프 등으로 *가설을 검정하는 과정*을 통해

주어진 데이터에 대해 알아보는 것.

### (1) EDA 대상(일변량/ 다변량)



**Univariate**(일변량)
EDA를 통해 한 번에 파악하려는 변수가 한 개
데이터를 설명하고 그 안에 존재하는 패턴을 찾는 것이 주 목적

**Multi-variate**(다변량)
파악하려는 변수가 여러 개
*여러 변수들 간의 관계*를 보는 것이 주 목적



### (2) EDA 종류 (시각화/ 비시각화)



**시각화**: 차트 혹은 그림 등을 이용



**비시각화**: 그래픽적인 요소 X, 주로 Summary Statistics 를 통해 데이터를 확인

### (3) EDA 유형

EDA 대상과 종류에 따라 EDA 유형이 구분됨

요인별 EDA 유형 구분

|  | 일반량 | 다변량 |
|------|------|------|
| 비시각화| 빈도표, 기술통계량 | 교차표, 상관계수 |
| 시각화 | 파이차트, 막대그래프, 히스토그램, 박스플롯 | 모자이크플롯, 박스플롯, 평행좌표, 산점도 |

- **Uni-Non Graphic(일변량 비시각화) :** 주어진 데이터의 Distribution을 확인하는 것이 주목적.
- **Uni-Graphic(일변량 시각화)** : 주어진 데이터를 전체적으로 살펴보는 것이 주목적.
- **Multi-Non Graphic(다변량 비시각화)** : 주어진 둘 이상의 변수 간 관계를 확인하는 것이 주목적.
- **Multi-Graphic(다변량 시각화)** : 주어진 둘 이상의 변수 간 관계를 전체적으로 살펴보는 것이 주목적.

## 3 전처리

(1) 결측값 처리  
  
변수의 결측값이 있는 상태로 데이터 분석 모델을 만들면 *변수 간의 관계가 왜곡*될 수 있음

데이터프레임 상에서 NaN, ?, 0 등의 값으로 나타나 있는 결측값을 확인  
- `isnull()` : 누락 데이터면 True, 유효한 데이터면 False 반환
    
    **→** `df.isnull().sum()` 의 형식으로 자주 사용!


- a. **삭제**: 결측값 데이터가 있는 행/ 열을 삭제  
삭제는 결측값이 무작위로 발생한 경우에 사용. 결측값이 무작위로 발생한 것이 아닌데 관측치를 삭제한 데이터를 사용할 경우, 왜곡된 모델이 생성될 수 있음.  
- b. **대체**: 결측값을 다른 값으로 대체
평균값, 최빈값 등의 대푯값을 활용 (fillna())

(2) 이상치 처리  
- describe() 통해 확인  
- 시각화 통해 확인(BoxPlot)  
- Z-score 통해 확인  
데이터를 쳥균(0)과 표준편차(1)로 정규화 후, 평균으로부터 얼마나 떨어져있는지 나타냄  
- Tukey Fences 통해 확인

**이상치 제거 및 처리 방법**   
a. 전체 삭제
- 이상값이 human error통해 발생한 경우엔 해당 관측치를 삭제하면 됨
- 단순 오타나, 주관식 설문 등의 비현실적 응답, 데이터 처리 과정에서의 오류 등의 경우에 사용

b. 다른 값으로 대체  
- 숫자가 작은 경우, 단순 삭제를 하면 *관측치의 절대령이 작아져* **신뢰성** 문제가 발생
- 뭔가 복잡하다 아무튼 대체해야 한다

c. 변수화  
- 이상값이 자연발생한 경우, 단순 삭제나 대체 X  
- 새로운 변수로 만들어버리기

d. 리샘플링
- 해당 이상값을 분리해서 모델 만들기




# 피처 엔지니어링 (변수 가공)

해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 피처(변수)들의 형태를 변형하거나 적절하게 처리하는 과정

**피처 = 입력 변수 = column**

변수 가공(피처 엔지니어링): 도메인 지식과 기존의 변수를 사용해서 기존의 데이터에 정보를 추가하는 일련의 과정

1. 레이블인코딩 vs 원핫인코딩

  
label encoding: 범주형 변수를 0부터 N-1까지의 숫자로 변환
one-hot encoding: 범주형 변수를 이진 벡터(0과 1)로 변환

2. 구간화 (binning)  
연속 데이터를 그대로 사용하기 보다는 일정한 구간으로 나눠서 분석하는 것이 효율적인 경우가 있음

3. 파생변수 생성  
기존의 피처를 다른 피처로 변환(재정의)하여 변수를 추가  
기존 데이터를 다른 관점에서 분석해 새로운 특성과 정보를 얻기 위한 과정  
정해진 원칙이 있는 방법이 아니어서 분석가의 도메인 지식과 데이터 특성에 대한 이해도에 따라 다양한 파생변수가 생성될 수 있음

4. 스케일링  
서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업  
왜곡 방지 위해 **정규화 작업**을 수행

정규화 작업: 각 열(변수)에 속하는 데이터 값을 동일한 크기 기준으로 나눈 비율로 나타내는 작업
