판다스의 개념: 데이터 분석에 쓰이는 파이썬 라이브러리로 행과 열을 다루게 해줍니다.

불러오는 방법: import pandas as pd 를 입력해주면 됩니다.

데이터프레임: 2차원 데이터 형식으로 하나 이상의 Series가 모여 있습니다.

데이터 확인하는 방법: 
    타입확인: type(Dataframe) 과 같은 예시처럼  type()의 괄호안에 타입을 확인하고 싶은 데이터를 넣으면 됩니다.

데이터 추출하는 방법:
    문자열에서 추출: 
        Indexing: str[0] = 첫번째 글자, 이처럼 하나의 글자를 뺴오는 것으로 안에 숫자로 순번을 정하면 음수가 들어있다면 거꾸로 순번을 셉니다
        Slicing: str[0:3] = 첫번째 글자부터 세번째 글자, 이처럼 특정 구간을 빼오는 것으로, :n 와 같이 앞이 없으면 처음부터, 반대로 n: 와 같이 뒤가 없으면 마지막까지 추출합니다
    리스트에서 추출
        Indexing: list[0] = 첫번째 값, 이처럼 하나의 값을 뺴오는 것으로 안에 순번을 정하는 법은 문자열에서 와 같습니다
        Slicing: list[0:3] = 첫번째 값부터 세번째 값, 이처럼 특정 구간의 값을 뺴오는 것으로 구간을 정하는 법은 문자열에서와 같습니다.
    튜플에서 추출 
        Indexing: tuple[0] = 첫번째 값, 이처럼 하나의 값을 뺴오는 것으로 안에 순번을 정하는 법은 문자열에서 와 같습니다
        Slicing: tuple[0:3] = 첫번째 값부터 세번째 값, 이처럼 특정 구간의 값을 뺴오는 것으로 구간을 정하는 법은 문자열에서와 같습니다.

데이터 조작하는 방법
    문자열
        덧셈이나 곱셈을 이용해 붙이거나 반복이 가능하고 count(),find 이외에 여러 함수들로 문자의 개수를 알거나 삽입,변경,제거를 할 수 있습니다.
    리스트
        리스트 간에도 문자열처럼 덧셈이나 곱셈이 가능하며 append,insert,remove 등을 통해 각각 추가하고,글자를 삽입하거나, 제거 할 수 있습니다.
    튜플
        한번 선언된 튜플은 변경이 불가능 하지만 덧셈과 곱셈은 가능합니다.
    세트
        집합의 형태로 저장되 있어 합집합,교집합,차집합과 같은 연산이 가능하며, .add 나 .update 와 같은 함수들로 집합내 데이터 조작이 가능합니다. 
    딕셔너리
        딕셔너리의 value 값을 리스트로 저장할수 있고 위에 리스트에서 한 방법으로 조작할 수 있습니다.
집계함수 사용하는 방법
    count() 몇개 있는지 셀 수 있습니다.
    sum() 합 을 구할수 있습니다.
    mean() 평균을 구할수 있습니다.
        이때 ()안에 axis=0,1을 넣어서 행이나 열의 합,또는 평균을 구할 수도 있습니다.
