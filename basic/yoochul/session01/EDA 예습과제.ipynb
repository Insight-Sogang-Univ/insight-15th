
EDA(탐색적 데이터 분석 과정) 
EDA는 수집한 데이터의 특징을 파악하는 것으로 데이터의 특성이나 패턴 그리고 데이터 간에 관계를 파악하며, 줭진 데이터를 탐색하고 이해하는 과정입니다.


데이터 파악 
데이터의 종류는 크게  범주형 그리고 수치형으로 나뉩니다. 그리고  읽는 형태로는 절대 파일이나 상대 경로를 통해 읽을 수 있으며 저희가 CSV파일이나 EXCEL 파일 같은 경우 판다스를 통해 데이터프레임 형식으로 읽을 수도 있습니다.

결측값 & 이상치 처리
먼저 결측값이란 데이터를 수집할때 측정되지 않거나 누락된 데이터를 뜻합니다.
NaN 이나 ?,0등의 값으로 나타나며 여러 함수들을 통해서 개수를 파악할수 있습니다.
또 함수들을 통해서 삭제할 수도 있으며,이 과정에서 정보가 손실될수 있으니 주의해야 합니다. 
다른 처리 방법은 대체가 있는데 평균이나, 최빈값 같이 표본을 대표할 수 있는 데이터로 결측값을 대체 합니다.
다음으로 이상치란 데이터의 범위에서 많이 벗어난 값으로 이역시 함수를 통해서 파악할 수 있으며, z점수를 활용해서 알 수도 있습니다.
처리방법은 결측값때와 같이 삭제 혹은 대체를 하거나 변수화,혹은 리샘플링을 통해 처리 할 수도 있습니다.

 피처 엔지니어링 (변수 가공)
 문제를 컴퓨터에 맞게 변형하는 것입니다. 방식은 총 4가지가 있으며 내용은 아래와 같습니다.
  레이블인코딩: 텍스트를 숫자로 바꿔주는 작업입니다.
  구간화: 일정한 구간으로 나눠서 분석하는 것입니다.
  파생변수 생성: 기존 데이터를 다른 관점으로 분석하기 위해서 재정의 하는것입니다.
  스케일링: 변수의 값 범위를 일정한 수준으로 맞춰주는 것입니다.

 데이터 시각화
 그래프나 차트,도표 같이 그림으로 변환해서 패턴,관계,추세를 파악하기 쉽게 하는 것으로 Matplotlib과 Seaborndmf 활용해 변환합니다.
