### 1) pandas
- 개념: 데이터 분석에 특화된 파이썬 라이브러리이며 행과 열로 이루어진 테이블 형태의 DataFrame을 기본 틀로 사용한다. Numpy의 확장판으로 구조화된 데이터 처리 기능을 보유한다.
- 라이브러리 불러오기: 
    1. import pandas as pd
    2. df=pd.read_csv('파일 경로')

### 2) dataframe
- 개념: 판다스에서 사용하는 주요 데이터 구조로, 컴퓨터로 정보를 정리하고 저장하는 도구
- 기본구조: 행(index)과 열(column)  

- Series: 값과 인덱스로 구성된 가장 간단한 1차원 자료구조
- DataFrame: 행과 열로 이루어진 2차원 배열 구조로, 여러 개의 시리즈가 모여 하나의 데이터프레임을 구성한다.

### 3) 데이터 확인
- df.head(n): 앞 n개의 행 반환 (기본값 n=5)
- df.tail(n): 뒤 n개의 행 반환 (기본값 n=5)  

- 데이터 정보 확인
    - df.shape: (행 개수, 열 개수) 반환
    - df.info(): 여러 정보들을 종합적으로 반환
    - df.describe(): 수치형 열들의 기본적인 통계값 반환
    - df.isnull(): 데이터에 결측치(NaN)가 있는지 여부 반환
    - df.dtypes: 열들의 데이터타입 반환

- 고유값 및 개수 확인
    - df.count(): 결측치의 개수를 제외한 데이터의 행 수 반환
    - df[열 이름].valus_counts(): 특정 열에 포함된 값들의 빈도수를 세어 반환
    - df[열 이름].unique(): 특정 열에 포함된 값들을 종류별로 1개씩만 반환


### 4) 데이터 추출
- 인덱싱(Indexing): 단일 항목을 지목할 때 사용
- 슬라이싱(Slicing): 범위를 지정해 여러 항목을 추출할 때 사용  

- 행단위 추출  
    - loc: 인자 지정 방법에 따라 행단위, 값단위, 열단위 추출 가능(끝 인덱스 포함하여 반환)
    - iloc: 인자로 정수형 인덱스 번호가 와야 한다. (끝 인덱스 포함하지 않고 반환)  

- 열단위 추출: df[열 이름], df[[열 이름]], df[[열 이름1, 열 이름2..]] 등으로 인덱싱  

- 불리언 인덱싱: 인덱스나 컬럼명이 아니라 특정 조건에 만족하는 데이터를 추출할 때 사용한다. True 또는 False의 배열을 통해 데이터를 필터링하는 연산이다.  

- 인덱스 설정: 0부터 시작하는 정수형 인덱스를 기본값으로 가지지만, 원하는 대로 변경 가능하다.


### 5) 데이터 조작
- 데이터 추가: df['새로운 열 이름'] = 리스트
- 데이터 삭제: df.drop(labels, axis)  

- 데이터 정렬
    - df.sort_index(): 인덱스를 기준으로 정렬
    - df.sort_values(by="열 이름"): 특정 열의 값을 기준으로 정렬

- 데이터 변환
    - df[열 이름] = df[열 이름].astype(바꿀 자료형)
    - df.replace(바꾸고 싶은 값, 바꿀 값)
    - df.transpose()
    - df.T

- 데이터프레임 병합
    - concat: 단순히 두 데이터프레임을 이어붙이는 방식
    - merge: 공통 항목을 기준으로 병합하는 방식, 형태가 다르더라도, 겹치는 데이터를 기준으로 데이터프레임을 합칠 수 있다.

### 6) 집계함수
- 개념: 여러 행으로부터 하나의 결과값을 반환하는 함수
    - `df['컬럼 명'].mean()` : 평균
    - `df['컬럼 명'].median()` : 중앙값
    - `df['컬럼 명'].std()` : 표준편차
    - `df['컬럼 명'].sum()` : 합계
    - `df['컬럼 명'].max()` : 최대값
    - `df['컬럼 명'].min()` : 최소값

- 그룹화(Grouping): 집계함수와 함께 사용해 일부 행을 대상으로 집계한다.
