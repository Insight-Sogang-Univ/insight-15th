### 1) EDA(Exploratory Data Analysis, 탐색적 데이터 분석 과정)
- 개념: 데이터 분석(모델링)을 위해 데이터를 여러 가지 방식으로 파악하는 모든 과정  
<br>
- 시각화 같은 도구를 통해서 패턴을 발견하거나, 데이터의 특이성을 확인하거나, 통계와 그래프로 가설을 검정하는 과정 등을 통해 주어진 데이터에 대해 알아보는 것  
<br>
(1) EDA 대상: 일변량 / 다변량  
(2) EDA 종류: 시각화 / 비시각화  
(3) EDA 유형: 일변량 비시각화 / 일변량 시각화 / 다변량 비시각화 / 다변량 시각화

### 2) 데이터셋 파악하기  
- 데이터 프레임 보기: head(n)  
<br>
- 데이터 변수 확인  
    (1) 변수 정의 확인: 어떤 정보를 가지는 변수인지 확인  
    (2) 변수 유형 확인: 질적/범주형과 양적/수치형으로 구분  
    (3) 변수 데이터 형식(자료형) 확인: 날짜, 수치, 텍스트 이미지 등의 구분 => df.info()  
    <br>
- 데이터 분포 확인: 단변수 분석 -> 원시 데이터의 평균값, 최빈값, 중간값 등 변수들의 분포를 산포도, 박스 플롯, 히스토그램 등의 그래프를 통해 하나의 데이터 분포를 확인해 분석할 수 있는 방법


### 3) 전처리
- 개념: 데이터를 분석에 적합한 형태로 만드는 과정  
<br>
- 결측값 처리  
    (1) 개념: 데이터 수집 과정에서 측정되지 않거나 누락된 데이터  
    <br>
    (2) 확인: 데이터 프레임 상에서 NaN, ?, 0 등의 값으로 나타나 있는 결측값 확인 (0 주의!)  
        - info(): 데이터 프레임의 요약 정보 출력
        - value_counts(dropna=False): 각 열의 결측값을 포함한 전체 데이터 확인 가능  
        - isnull(): 누락 데이터면 True, 유효한 데이터면 False 반환  
        - notnull(): 유효 데이터면 True, 누락 데이터면 False 반환  
        - replace(): 0이나 ? 등을 NaN으로 변환 가능  
    <br>
    (3) 처리  
        - 삭제: 데이터가 있는 행 또는 열을 삭제 => dropna()  
        - 대체: 다른 값으로 대체 (평균값, 최빈값 등의 대푯값 활용)  
<br>
- 이상치 처리  
    (1) 개념: 관측된 데이터의 범위에서 많이 벗어난 값  
    <br>
    (2) 확인  
        - 통계를 통해 확인: describe()  
        - 시각화를 통해 확인: BoxPlot  
        - Z-score를 통해 확인: 데이터를 평균과 표준편차로 정규화하여, 평균으로부터 얼마나 떨어져있는지를 나타냄 => 일반적으로 2~3을 넘어가는 데이터를 이상치로 간주  
        - Tukey Fences를 통해 확인: 사분위 범위(IQR)을 기반으로 Q1 - (1.5 * IQR) 미만, Q3 + (1.5 * IQR) 초과  
    <br>
    (3) 제거 및 처리  
        - 전체 삭제  
        - 다른 값으로 대체  
        - 변수화  
        - 리샘플링  

### 4) 피처 엔지니어링 (변수 가공)  
- 개념: 해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 피처(변수)들의 형태를 변형하거나 적절하게 처리하는 과정  
- 도메인 지식과 기존의 변수를 사용해서 기존의 데이터에 정보를 추가하는 일련의 과정  
<br>
    - Label Encoding: 범주형 변수를 0부터 N-1까지의 숫자로 변환  
    - One-Hot Encoding: 범주형 변수를 이진 벡터로 변환  
    - 구간화(Binning): 연속 데이터를 그대로 사용하기보다는 일정한 구간으로 나눠서 분석  
    - 파생변수 생성: 기존의 피처를 다른 피처를 변환(재정의)하여 변수 추가  
    - 스케일링: 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업 => StandardScaler, MinMaxScaler  

### 5) 데이터 시각화  
- 개념: 데이터에 대한 이해를 돕기 위해 그림, 도형 등 그래픽 요소들을 이용해 데이터를 묘사하고 표현하는 과정  
    - Matplotlib  
    - Seaborn  
<br>
- 파라미터: 프로그래밍된 함수의 입력값  
<br>
- 그래프 종류  
    - boxplot(상자 수염 그림): 사분위수와 이상치를 보여주는 그래프  
    - countplot: 범주형 변수의 빈도수를 확인하는 그래프  
    - hisplot: 도수분포표를 그래프로 나타낸 것, 수치형 변수의 구간별 빈도수를 보여줌  
    - displot, kdeplot(커널밀도추정 그래프): 히스토그램을 연속적으로 곡선으로 연결한 그래프  
    - barplot: 범주형 데이터 값 x에 따른 수치형 데이터 값 y의 평균 값을 제공  
    - pointplot: barplot과 모양만 다르고 동일한 정보 제공  
    - scatterplot(산점도 그래프): 두 연속형 변수 간의 관계를 시각화하기 위해 사용되는 그래프, 각각의 데이터 포인트는 두 변수의 값을 나타내며, x축과 y축에 데이터 포인트를 분산하여 그림  
    - regplot(회귀선이 추가된 산점도 그래프): 두 개의 연속 변수 사이의 산점도를 그리고 회귀선을 함께 나타내는 그래프  
    - catplot: 수치형 데이터와 범주형 데이터의 관계를 볼 때 사용  
    - pieplot: 데이터의 부분과 전체 간의 비율을 표현하는 그래프  
    - heatmap: 변수간 상관계수를 직관적으로 볼 수 있는 그래프  
    - violinplot: 박스 플롯과 커널밀도추정 함수 그래프를 합친 그래프  
    - pairplot: 여러 그래프를 한번에 설정
